Ici le dérive la chimie pour faire un essai en 1D et voir si j'ai la bonne forme.

L'équation de dissolution d'un mineral traceur et de l'évolution des concentrations s'écrit:

- En version explicite

$$
  \left\{
    \begin{array}{l}
      \overline{c}_{i}^{2n+1} = \overline{c}_{i}^{2n}
      \\
      \mathcal{A}_{i} \omega_{i} \dfrac{c_{i}^{2n+1} - c_{i}^{2n}}{\Delta t^{2n}} = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \big(D_{e, ij}\dfrac{c_{j}^{2n+1} - c_{i}^{2n+1}}{ \left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\big)
      \\
      \mathcal{A}_{i} \omega_{i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n}} = \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right)
      \\
      \mathcal{A}_{i} \omega_{i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n}} = - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right)
    \end{array}
  \right.
$$

with the fix point criterion: $$ \left(1 - \dfrac{c_{i}^{2n+2}}{c_{i}^{2n}} \right) < \epsilon $$

$$ \mathcal{L} = \dfrac{1}{2} \sum_{n} \sum_{i} \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{\sigma_{c_{i}^{n, obs}}}\right)^{2} - \langle \lambda_{c_{i}^{n+1}}, g_{c} \rangle - \langle \lambda_{\overline{c}_{i}^{n+1}}, g_{\overline{c}} \rangle $$

Derivation par $\overline{c}_{i}^{2n}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n}} \langle \lambda_{\overline{c}_{i}}^{2n+1}, \overline{c}_{i}^{2n+1} - \overline{c}_{i}^{2n} \rangle$$

$$ = \lambda_{\overline{c}_{i}}^{2n} - \lambda_{\overline{c}_{i}}^{2n+1}$$

Derivation par $\overline{c}_{i}^{2n+1}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n+1}} \biggl< \lambda_{\overline{c}_{i}}^{2n+2}, \mathcal{A}_{i} \omega_{i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \lambda_{\overline{c}_{i}}^{2n+2} \Bigg( \dfrac{-\mathcal{A}_{i} \omega_{i}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \Bigg) + \lambda_{\overline{c}_{i}}^{2n+1} \Bigg( \dfrac{\mathcal{A}_{i} \omega_{i}}{\Delta t^{2n-1}}\Bigg) $$

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n+1}} \biggl< \lambda_{c_{i}}^{2n+2}, \mathcal{A}_{i} \omega_{i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n}} + \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \lambda_{c_{i}}^{2n+2} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right)$$

On obtient alors une expression explicite pour les teneurs adjointes:

$$ \lambda_{\overline{c}_{i}}^{2n+1} = \Bigg( \lambda_{\overline{c}_{i}}^{2n+2} \left( \dfrac{1}{\Delta t^{2n}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right) - \lambda_{c_{i}}^{2n+2} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \Bigg) \Delta t^{2n-1} $$

2) Derivation par $c_{i}^{2n}$:


$$ \dfrac{\partial}{\partial c_{i}^{2n}} \biggl< \lambda_{c_{i}}^{2n+1}, \mathcal{A}_{i} \omega_{i} \dfrac{c_{i}^{2n+1} - c_{i}^{2n}}{\Delta t^{2n}} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \big(D_{e, ij}\dfrac{c_{j}^{2n+1} - c_{i}^{2n+1}}{ \left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\big) \biggr>$$

$$= \mathcal{A}_{i} \omega_{i} \Big( \dfrac{\lambda_{c_{i}}^{2n}}{\Delta t^{2n-1}} - \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n}} \Big) - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \left( D_{e, ij}\dfrac{\lambda_{c_{j}}^{2n} - \lambda_{c_{i}}^{2n}}{ \left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\right) $$

$$ \dfrac{\partial}{\partial c_{i}^{2n}} \dfrac{1}{2} \sum_{n} \sum_{i} \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{\sigma_{c_{i}^{2n, obs}}}\right)^{2}$$

$$ = - \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\sigma_{c_{i}^{2n, obs}}}^{2}}\right)$$

3) Derivation par $c_{i}^{2n+1}$:

$$ \dfrac{\partial}{\partial c_{i}^{2n+1}} \biggl< \lambda_{\overline{c}_{i}}^{2n+2}, \mathcal{A}_{i} \omega_{i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \lambda_{\overline{c}_{i}}^{2n+2} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}} $$

$$ \dfrac{\partial}{\partial c_{i}^{2n+1}} \biggl< \lambda_{c_{i}}^{2n+2}, \mathcal{A}_{i} \omega_{i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n}} + \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \mathcal{A}_{i} \omega_{i}  \left(\dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n-1}} -  \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} \right) - \lambda_{c_{i}}^{2n+2} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}}$$

Gathering the three previous, it yields

$$\dfrac{\partial \mathcal{L}}{\partial c_{i}^{2n+1}} = - \mathcal{A}_{i} \omega_{i}  \left( \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n-1}} \right) + \mathcal{A}_{i} \omega_{i} \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} \right) + \lambda_{c_{i}}^{2n+2} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}}  - \lambda_{\overline{c}_{i}}^{2n+2} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}} - \left( \dfrac{c_{i}^{2n+2, obs} - c_{i}^{2n+2, calc}}{{\sigma_{c_{i}^{2n=2, obs}}}^{2}}\right) = 0 $$

$$ \implies \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} + \left(\lambda_{c_{i}}^{2n+2} - \lambda_{\overline{c}_{i}}^{2n+2} \right) k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}} - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\mathcal{A}_{i} \omega_{i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n-1} $$

Derivation du point fixe:

$$ \dfrac{\partial}{\partial c_{i}^{2n}} \langle \lambda_{c_{i}}^{2n+2}, \left(1 - \dfrac{c_{i}^{2n+2}}{c_{i}^{2n}} \right) - \epsilon \rangle$$


$$ \implies \dfrac{\lambda_{c_{i}}^{2n+2}}{{c_{i}^{2n}}^{2}} - \lambda_{c_{i}}^{2n} < 0 $$

Dérivation par rapport à $\overline{c}_{i}^{0}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \langle \lambda_{\overline{c}_{i}}^{1}, \overline{c}_{i}^{1} - \overline{c}_{i}^{0} \rangle = \lambda_{\overline{c}_{i}}^{0} - \lambda_{\overline{c}_{i}}^{1} = 0$$

Dérivation par rapport à $\overline{c}_{i}^{0}$:

$$ \dfrac{\partial \mathcal{L}}{\partial \overline{c}_{i}^{0}} = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \biggl< \lambda_{\overline{c}_{i}^{2n+2}}, \mathcal{A}_{i} \omega{i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \sum_{n=0}^{N} \lambda_{\overline{c}_{i}^{2n+2}} \left( \mathcal{A}_{i} \omega{i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right)$$

Et, en remarquant que la boucle directe sur les temps commence à $n=1$, alors $\overline{c}_{i}^{1} = \overline{c}_{i}^{0}$, et $c_{i}^{1} = c_{i}^{0}$ alors la dérivée est non nulle pour $n=0$ uniquement:

$$ \dfrac{\partial \mathcal{L}}{\partial \overline{c}_{i}^{0}} = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \left[\lambda_{\overline{c}_{i}^{2}} \left( \mathcal{A}_{i} \omega{i} \dfrac{\overline{c}_{i}^{2} - \overline{c}_{i}^{0}}{\Delta t^{0}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{0} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right) \right] $$

$$ = \lambda_{\overline{c}_{i}^{2}} \left( \mathcal{A}_{i} \omega{i} \dfrac{- 1}{\Delta t^{0}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s}  \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right) $$



Et de la même manière:

$$ \dfrac{\partial \mathcal{L}}{\partial \overline{c}_{i}^{0}} = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \biggl< \lambda_{c_{i}^{2n+2}}, \mathcal{A}_{i} \omega{i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n}} + \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \biggr>$$

$$ = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \sum_{n=0}^{N} \lambda_{c_{i}^{2n+2}} \left( \mathcal{A}_{i} \omega{i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right)$$

$$ = \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \left[\lambda_{c_{i}^{2}} \left( \mathcal{A}_{i} \omega{i} \dfrac{c_{i}^{2} - c_{i}^{0}}{\Delta t^{0}} - \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \overline{c}_{i}^{0} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right) \right] $$

$$ = \lambda_{c_{i}^{2}} \mathcal{A}_{i} \omega_{i} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) $$

$$\implies \dfrac{d J}{d {\overline{c}_{i}}^{0}} = \mathcal{A}_{i} \omega_{i} \Bigg[ {\lambda_{\overline{c}_{i}}^{2}} \left(\dfrac{1}{\Delta t^{0}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right) - \lambda_{c_{i}^{2}} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \Bigg]$$

$$
  \left\{
    \begin{array}{l}
      \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} -  \lambda_{\overline{c}_{i}}^{2n+2} \dfrac{k_{v} A_{s} \overline{c}_{i}^{2n+1}}{K_{s}}  - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\mathcal{A}_{i} \omega_{i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n-1}
      \\
      \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} + \left(\lambda_{c_{i}}^{2n+2} - \lambda_{\overline{c}_{i}}^{2n+2} \right) k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}} - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\mathcal{A}_{i} \omega_{i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n-1}
      \\
      \mathcal{A}_{i} \omega_{i} \Big( \dfrac{\lambda_{c_{i}}^{2n}}{\Delta t^{2n-1}} - \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n}} \Big) - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \left( D_{e, ij}\dfrac{\lambda_{c_{j}}^{2n} - \lambda_{c_{i}}^{2n}}{ \left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\right)
      \\
      \lambda_{\overline{c}_{i}}^{2n} = \lambda_{\overline{c}_{i}}^{2n+1}
    \end{array}
  \right.
$$

```python
from typing import List, Tuple, Optional
import numpy as np
import matplotlib.pyplot as plt
import numpy.typing as npt
import pyrtid.forward as dmfwd
from pyrtid.forward.models import (
    TransportModel,
    get_owner_neigh_indices,
    TransportParameters,
    GeochemicalParameters,
    Geometry,
    TimeParameters,
)
from pyrtid.inverse.adjoint.amodels import AdjointTransportModel
from pyrtid.utils import harmonic_mean, dxi_harmonic_mean, get_super_lu_preconditioner
```

```python
NDArrayFloat = npt.NDArray[np.float_]
```

```python
nx = 20  # number of voxels along the x axis
ny = 1  # number of voxels along the y axis
dx = 2.98  # voxel dimension along the x axis
dy = 3.7  # voxel dimension along the y axis
dz = 1

geometry = Geometry(nx, ny, dx, dy, dz)

time_params = TimeParameters(
    nt=50,  # number of time steps
    dt_init=500000.0,  # timestep in seconds
    # dt_min=500.0,
    # dt_max=5000000.0,
)

# Chemistry parameters
c0 = 0.0001  # general initial concentration [molal]
M0 = 0.001  # mineral grade [mol/kg] -> kg of water
M1 = 0.0005  # mineral grade [mol/kg] -> kg of water
M1 = np.abs(0.001 * (0.5 + 0.5 * np.sin(np.arange(nx) / 2))).reshape(20, -1)
kv = -6.9e-9  # kinetic rate,       [mol/m2/s]
moleweight = 270.0  # molar weight [g/mol]
surface = 500  # cm2/g
As = moleweight * surface / 1e4  # specific area [m2/mol]
logK = 3.2
Ks = 1.0 / pow(10, logK)  # solubility constant [no unit]
porosity = 0.23
diffusion = 1.0 - 4

gch_params_ref = GeochemicalParameters(c0, M0, kv, As, Ks)

gch_params_est = GeochemicalParameters(c0, M0, kv, As, Ks)

tr_params = TransportParameters(diffusion=diffusion, porosity=porosity)
```

```python
plt.plot(M1)
```

```python
from scipy.sparse import csc_matrix, lil_matrix
from scipy.sparse.linalg import LinearOperator, splu, gmres


def make_transport_matrices_diffusion_only(
    geometry: dmfwd.Geometry,
    tr_model: TransportModel,
    time_params: dmfwd.TimeParameters,
) -> Tuple[csc_matrix, csc_matrix]:
    """
    Make matrices for the transport.

    Note
    ----
    Since the diffusion coefficient does not vary with time,
    matrices q_prev and q_next are the same.
    """

    dim = geometry.nx * geometry.ny
    q_prev = lil_matrix((dim, dim), dtype=np.float64)
    q_next = lil_matrix((dim, dim), dtype=np.float64)

    # X contribution
    if geometry.nx >= 2:
        dmean: NDArrayFloat = np.zeros((geometry.nx, geometry.ny), dtype=np.float64)
        dmean[:-1, :] = harmonic_mean(
            tr_model.effective_diffusion[:-1, :], tr_model.effective_diffusion[1:, :]
        )
        dmean = dmean.flatten(order="F")

        # Forward scheme:
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(0, geometry.nx - 1), slice(None)),
            (slice(1, geometry.nx), slice(None)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dx**2
        )
        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dx**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dx**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dx**2
        )

        # Backward scheme
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(1, geometry.nx), slice(None)),
            (slice(0, geometry.nx - 1), slice(None)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dx**2
        )
        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dx**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dx**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dx**2
        )

    # Y contribution
    if geometry.ny >= 2:
        dmean: NDArrayFloat = np.zeros((geometry.nx, geometry.ny), dtype=np.float64)
        dmean[:, :-1] = harmonic_mean(
            tr_model.effective_diffusion[:, :-1], tr_model.effective_diffusion[:, 1:]
        )
        dmean = dmean.flatten(order="F")

        # Forward scheme:
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(None), slice(0, geometry.ny - 1)),
            (slice(None), slice(1, geometry.ny)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dy**2
        )
        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dy**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dy**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dy**2
        )

        # Backward scheme
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(None), slice(1, geometry.ny)),
            (slice(None), slice(0, geometry.ny - 1)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dy**2
        )
        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dy**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dy**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dy**2
        )

    # Add 1/dt for the left term contribution
    q_next.setdiag(q_next.diagonal() + tr_model.porosity.flatten("F") / time_params.dt)
    q_prev.setdiag(q_prev.diagonal() + tr_model.porosity.flatten("F") / time_params.dt)

    return q_next.tocsc(), q_prev.tocsc()


def solve_transport(
    tr_model: TransportModel,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the transport problem."""
    q_next, q_prev = make_transport_matrices_diffusion_only(
        geometry, tr_model, time_params
    )

    # Multiply prev matrix by prev vector
    tmp = q_prev.dot(tr_model.conc[:, :, time_index - 1].flatten(order="F"))

    # Build the LU preconditioning
    preconditioner = get_super_lu_preconditioner(q_next)

    # Solve Ax = b with A sparse using LU preconditioner
    res, exit_code = gmres(
        q_next, tmp, x0=None, M=preconditioner, atol=tr_model.tolerance
    )

    # In that regard, we save the intermediate concentrations for the non
    # iterative sequential apprach (adjoint state)
    tr_model.conc_post_tr[:, :, time_index] = res.reshape(geometry.ny, geometry.nx).T

    return exit_code


def solve_geochem_explicit(
    tr_model: TransportModel,
    gch_params: GeochemicalParameters,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the geochemistry part."""

    # Need to take into account boundary conditions:
    # And then the reactive (chemistry) contribution with the updated conc

    # Note: the concentration has been modified by the transport so we need to use the
    # updated one.
    m0 = tr_model.grade[:, :, time_index - 1]

    dMdt = (
        gch_params.kv
        * gch_params.As
        * m0
        * (1.0 - tr_model.conc_post_tr[:, :, time_index - 1] / gch_params.Ks)
    )

    tr_model.grade[:, :, time_index] = m0 + dMdt * time_params.dt
    tr_model.conc[:, :, time_index] = (
        tr_model.conc_post_tr[:, :, time_index] - dMdt * time_params.dt
    )
```

```python
tr_model_ref = TransportModel(geometry, time_params, tr_params, gch_params_ref)
tr_model_ref.grade[:, :, 0] = M0
tr_model_est = TransportModel(geometry, time_params, tr_params, gch_params_est)
tr_model_est.grade[:, :, 0] = M1

dt_list = [0.0]
time_params.reset_dt()
for it in range(1, time_params.nt + 1):
    time_params.update_dt(1)  # we consider only one iteration.
    solve_transport(tr_model_ref, time_params, it)
    solve_geochem_explicit(tr_model_ref, gch_params_ref, time_params, it)

time_params.reset_dt()
for it in range(1, time_params.nt + 1):
    time_params.update_dt(1)  # we consider only one iteration.
    solve_transport(tr_model_est, time_params, it)
    solve_geochem_explicit(tr_model_est, gch_params_est, time_params, it)
```

$$
  \left\{
    \begin{array}{l}
      \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n}} + \left(\lambda_{c_{i}}^{2n+2} - \lambda_{\overline{c}_{i}}^{2n+2} \right) k_{v} A_{s} \dfrac{\overline{c}_{i}^{2n+1}}{K_{s}} - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\mathcal{A}_{i} \omega_{i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n-1}
      \\
      \lambda_{\overline{c}_{i}}^{2n+1} = \Bigg( \lambda_{\overline{c}_{i}}^{2n+2} \left( \dfrac{1}{\Delta t^{2n}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right) - \lambda_{c_{i}}^{2n+2} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \Bigg) \Delta t^{2n-1} 
      \\
      \mathcal{A}_{i} \omega_{i} \Big( \dfrac{\lambda_{c_{i}}^{2n}}{\Delta t^{2n-1}} - \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n}} \Big) - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \left( D_{e, ij}\dfrac{\lambda_{c_{j}}^{2n} - \lambda_{c_{i}}^{2n}}{ \left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\right)
      \\
      \lambda_{\overline{c}_{i}}^{2n} = \lambda_{\overline{c}_{i}}^{2n+1}
    \end{array}
  \right.
$$

```python
def add_chem_adjoint_sources(
    tr_model: TransportModel,
    a_tr_model: AdjointTransportModel,
    geometry: Geometry,
    a_sources: NDArrayFloat,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Add the sources in the adjoint problem."""
    # Note the sources are added on the previous timestep
    # -> this is the time at which the observation has been made !
    # Skip the last timestep (there is no chemistry nor transport between n=0 and n=2)
    # if time_index == 0:
    #     return
    dt_cur = time_params.ldt[time_index]
    a_tr_model.a_conc[:, :, time_index + 1] -= (
        a_sources[:, :, time_index + 1]
        / geometry.mesh_area
        / tr_model.porosity
        * dt_cur
    )


def solve_adj_geochem_explicit(
    tr_model: TransportModel,
    a_tr_model: AdjointTransportModel,
    gch_params: GeochemicalParameters,
    geometry: Geometry,
    a_sources,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the geochemistry part."""

    # 1) Add the adjoint sources on the previous timestep
    add_chem_adjoint_sources(
        tr_model_est, a_tr_model, geometry, a_sources, time_params, it
    )

    # # Skip the last timestep (there is no chemistry nor transport between n=0 and n=1)
    if time_index == 0:
        return

    # Adjoint variable at the previous timestep going backward (timestep n+1)
    am_prev = a_tr_model.a_grade[:, :, time_index + 1]
    ac_prev = a_tr_model.a_conc[:, :, time_index + 1]
    dt_prev = time_params.ldt[time_index]
    # Forward variable at the current timestep going backward
    # (timestep n -> we want to predict)
    m_cur = tr_model.grade[:, :, time_index]
    c_prev_post_tr = tr_model.conc_post_tr[:, :, time_index + 1]  # 2n + 1
    dt_cur = time_params.ldt[time_index - 1]

    # Update concentration (with adjoint sources)
    a_tr_model.a_conc_post_gch[:, :, time_index] = (
        ac_prev / dt_prev
        + (ac_prev - am_prev) * gch_params.kv * gch_params.As * m_cur / gch_params.Ks
    ) * dt_cur

    # Update mineral value
    a_tr_model.a_grade_post_gch[:, :, time_index] = (
        am_prev
        * (
            1 / dt_prev
            + gch_params.kv * gch_params.As * (1.0 - c_prev_post_tr / gch_params.Ks)
        )
        # - a_tr_model.a_conc_post_gch[:, :, time_index] / dt_cur + ac_prev / dt_prev
        - ac_prev * gch_params.kv * gch_params.As * (1 - c_prev_post_tr / gch_params.Ks)
    ) * dt_cur
```

```python
def make_transient_adj_transport_matrices(
    geometry: Geometry, tr_model: TransportModel, time_params: TimeParameters
) -> Tuple[csc_matrix, csc_matrix]:
    """
    Make matrices for the transient transport.

    Note
    ----
    Since the diffusion coefficient and porosity does not vary with time,
    matrices q_prev and q_next are the same.
    """

    dim = geometry.nx * geometry.ny
    q_prev = lil_matrix((dim, dim), dtype=np.float64)
    q_next = lil_matrix((dim, dim), dtype=np.float64)

    # X contribution
    dmean: NDArrayFloat = np.zeros((geometry.nx, geometry.ny), dtype=np.float64)
    dmean[:-1, :] = harmonic_mean(
        tr_model.effective_diffusion[:-1, :], tr_model.effective_diffusion[1:, :]
    )
    dmean = dmean.flatten(order="F")

    # Forward scheme:
    idc_owner, idc_neigh = get_owner_neigh_indices(
        geometry,
        (slice(0, geometry.nx - 1), slice(None)),
        (slice(1, geometry.nx), slice(None)),
        tr_model.cst_conc_indices,
    )

    q_next[idc_owner, idc_neigh] -= (
        tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dx**2
    )
    q_next[idc_owner, idc_owner] += (
        tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dx**2
    )
    q_prev[idc_owner, idc_neigh] += (
        (1.0 - tr_model.crank_nicolson_diffusion) * dmean[idc_owner] / geometry.dx**2
    )
    q_prev[idc_owner, idc_owner] -= (
        (1.0 - tr_model.crank_nicolson_diffusion) * dmean[idc_owner] / geometry.dx**2
    )

    # Backward scheme
    idc_owner, idc_neigh = get_owner_neigh_indices(
        geometry,
        (slice(1, geometry.nx), slice(None)),
        (slice(0, geometry.nx - 1), slice(None)),
        tr_model.cst_conc_indices,
    )

    q_next[idc_owner, idc_neigh] -= (
        tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dx**2
    )
    q_next[idc_owner, idc_owner] += (
        tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dx**2
    )
    q_prev[idc_owner, idc_neigh] += (
        (1.0 - tr_model.crank_nicolson_diffusion) * dmean[idc_neigh] / geometry.dx**2
    )
    q_prev[idc_owner, idc_owner] -= (
        (1.0 - tr_model.crank_nicolson_diffusion) * dmean[idc_neigh] / geometry.dx**2
    )

    # Y contribution
    if geometry.ny >= 2:
        dmean: NDArrayFloat = np.zeros((geometry.nx, geometry.ny), dtype=np.float64)
        dmean[:, :-1] = harmonic_mean(
            tr_model.effective_diffusion[:, :-1], tr_model.effective_diffusion[:, 1:]
        )
        dmean = dmean.flatten(order="F")

        # Forward scheme:
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(None), slice(0, geometry.ny - 1)),
            (slice(None), slice(1, geometry.ny)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dy**2
        )
        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_owner] / geometry.dy**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dy**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_owner]
            / geometry.dy**2
        )

        # Backward scheme
        idc_owner, idc_neigh = get_owner_neigh_indices(
            geometry,
            (slice(None), slice(1, geometry.ny)),
            (slice(None), slice(0, geometry.ny - 1)),
            tr_model.cst_conc_indices,
        )

        q_next[idc_owner, idc_neigh] -= (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dy**2
        )
        q_next[idc_owner, idc_owner] += (
            tr_model.crank_nicolson_diffusion * dmean[idc_neigh] / geometry.dy**2
        )
        q_prev[idc_owner, idc_neigh] += (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dy**2
        )
        q_prev[idc_owner, idc_owner] -= (
            (1.0 - tr_model.crank_nicolson_diffusion)
            * dmean[idc_neigh]
            / geometry.dy**2
        )

    # Add 1/dt for the left term contribution
    q_next.setdiag(q_next.diagonal() + tr_model.porosity.flatten("F") / time_params.dt)
    q_prev.setdiag(q_prev.diagonal() + tr_model.porosity.flatten("F") / time_params.dt)

    return q_next.tocsc(), q_prev.tocsc()


def solve_adj_transport_transient_semi_implicit(
    geometry: Geometry,
    tr_model: TransportModel,
    a_tr_model: AdjointTransportModel,
    time_params: TimeParameters,
    time_index: int,
) -> int:
    """
    Solving the adjoint diffusivity equation:

    dc/dt = div D grad c + ...
    """

    # Skip the last timestep (there is no transport between n=0 and n=1)
    if time_index == 0:
        return 0

    q_next, q_prev = make_transient_adj_transport_matrices(
        geometry, tr_model, time_params
    )

    # Get the previous vector
    prev_vector = a_tr_model.a_conc_post_gch[:, :, time_index].ravel("F")

    # Multiply prev matrix by prev vector
    tmp = q_prev.dot(prev_vector)

    # Build the LU preconditioning
    preconditioner = get_super_lu_preconditioner(q_next)

    # Solve Ax = b with A sparse using LU preconditioner
    res, exit_code = gmres(q_next, tmp, M=preconditioner, atol=1e-15)
    # Note: we go backward in time, so time_index -1...
    a_tr_model.a_conc[:, :, time_index] = res.reshape(geometry.ny, geometry.nx).T

    a_tr_model.a_grade[:, :, time_index] = a_tr_model.a_grade_post_gch[:, :, time_index]

    return exit_code


# def solve_adj_transport(
#     tr_model: TransportModel,
#     a_tr_model: AdjointTransportModel,
#     gch_params: GeochemicalParameters,
#     geometry: Geometry,
#     time_params: TimeParameters,
#     time_index: int,
# ) -> None:
#     # Skip the last timestep (there is no transport between n=0 and n=1)
#     if time_index == 0:
#         return
#     a_tr_model.a_conc[:, :, time_index] = a_tr_model.a_conc_post_gch[
#         :, :, time_index
#     ] * (1 + tr_model.diffusion)
#     a_tr_model.a_grade[:, :, time_index] = a_tr_model.a_grade_post_gch[:, :, time_index]
```

```python
gch_params = gch_params_est


def get_gradient(time_params: TimeParameters):
    return (
        (
            atr_model.a_grade[:, :, 1]
            * (
                1 / time_params.ldt[0]
                + gch_params.kv
                * gch_params.As
                * (1.0 - tr_model_est.conc[:, :, 0] / gch_params.Ks)
            )
            - atr_model.a_conc[:, :, 1]
            * gch_params.kv
            * gch_params.As
            * (1 - tr_model_est.conc[:, :, 0] / gch_params.Ks)
        )
        * geometry.mesh_area
        * porosity
    )


adjoint_gradient = get_gradient(time_params=time_params).ravel()
```

```python
def objfun(x, time_params: TimeParameters) -> float:
    _tr_model = TransportModel(geometry, time_params, tr_params, gch_params_est)
    _tr_model.grade[:, :, 0] = x.reshape(nx, ny, order="F")

    time_params.reset_dt()
    for it in range(1, time_params.nt + 1):
        time_params.update_dt(1)
        solve_transport(_tr_model, time_params, it)
        solve_geochem_explicit(_tr_model, gch_params_est, time_params, it)

    a_sources = tr_model_ref.conc - _tr_model.conc
    return float(0.5 * np.sum(a_sources**2))
```
