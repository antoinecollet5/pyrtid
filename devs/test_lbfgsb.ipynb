# import some functions from stochopy to test the code

```python
# python lbfgsb works with numpy arrays
import numpy as np
import pyrtid.inverse.lbfgsb as lbfgsb
from scipy.optimize import minimize
```

# First example: min x^Tx such that x>=1. Optimal solution hits the boundary.

```python
def quad(x):
    return x.dot(x)


def grad_quad(x):
    return x


ftol = 1e-5
gtol = 1e-5
l = np.array([1, 1])
u = np.array([np.inf, np.inf])
bounds = np.array((l, u)).T
x0 = np.array([5, 5])
```

```python
bounds[0]
```

### New implementation

```python
print("====================== Quadratic example ======================")
opt_quad = lbfgsb.L_BFGS_B(
    x0=x0, fun=quad, jac=grad_quad, bounds=bounds, ftol=ftol, gtol=gtol
)
print("")
print("")
xOpt = np.array([1, 1])
theoOpt = {"x": xOpt, "f": quad(xOpt), "df": grad_quad(xOpt)}
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
print(opt_quad)
```

### Scipy

```python
bounds
```

```python
minimize(
    quad,
    x0,
    jac=grad_quad,
    bounds=bounds,
    method="l-bfgs-b",
    options={"gtol": gtol, "ftol": ftol},
)
```

## Second example : min Rosenbrock function

```python
def rosenbrock(x):
    return 100 * (x[1] - x[0] ** 2) ** 2 + (1 - x[0]) ** 2


def grad_rosenbrock(x):
    g = np.empty(x.size)
    g[0] = 400 * x[0] * (x[0] ** 2 - x[1]) + 2 * (x[0] - 1)
    g[1] = 200 * (-x[0] ** 2 + x[1])
    return g


l = np.array([-2, -2])
u = np.array([2, 2])
bounds = np.array((l, u)).T
# x0 = np.array([0.12, 0.12])
x0 = np.array([-1, -1])
```

### New implementation

```python
np.finfo(1.0).resolution * 1e7
```

```python
opt_rosenbrock = lbfgsb.L_BFGS_B(
    x0=x0, fun=rosenbrock, jac=grad_rosenbrock, bounds=bounds, ftol=ftol, gtol=gtol
)
print("")
print("")
theoOpt = {"x": np.array([1, 1]), "f": 0, "df": grad_rosenbrock(np.array([1, 1]))}
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
print(opt_rosenbrock)
```

### Scipy

```python
minimize(
    rosenbrock,
    x0,
    jac=grad_rosenbrock,
    bounds=bounds,
    method="l-bfgs-b",
    options={"gtol": gtol, "ftol": ftol},
)
```

## Third example : min Beale function

```python
def beale(x):
    return (
        (1.5 - x[0] + x[0] * x[1]) ** 2
        + (2.25 - x[0] + x[0] * x[1] ** 2) ** 2
        + (2.625 - x[0] + x[0] * x[1] ** 3) ** 2
    )


def grad_beale(x):
    y1 = x[1]
    y2 = y1 * y1
    y3 = y2 * y1
    f1 = 1.5 - x[0] + x[0] * y1
    f2 = 2.25 - x[0] + x[0] * y2
    f3 = 2.625 - x[0] + x[0] * y3

    return np.array(
        [
            2 * (y1 - 1) * f1 + 2 * (y2 - 1) * f2 + 2 * (y3 - 1) * f3,
            2 * x[0] * f1 + 4 * x[0] * y1 * f2 + 6 * x[0] * y2 * f3,
        ]
    )


ftol = 1e-14
gtol = 1e-10
l = -4.5 * np.ones(2)
u = -l
bounds = np.array((l, u)).T
x0 = np.array([2.5, -1.3])
```

### New implementation

```python
opt_beale = lbfgsb.L_BFGS_B(
    x0=x0, fun=beale, jac=grad_beale, bounds=bounds, ftol=ftol, gtol=gtol
)
print("")
print("")
theoOpt = {
    "x": np.array([3, 0.5]),
    "f": beale(np.array([3, 0.5])),
    "df": grad_beale(np.array([3, 0.5])),
}
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
print(opt_beale)
```

### Scipy

```python
minimize(
    beale,
    x0,
    jac=grad_beale,
    bounds=bounds,
    method="l-bfgs-b",
    options={"gtol": gtol, "ftol": ftol},
)
```
