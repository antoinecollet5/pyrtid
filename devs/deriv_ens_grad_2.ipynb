# Derivation of an itreative ensemble smoother using a gradient of the objective function with the adjoint state

The objective function reads:

\begin{equation}
\begin{aligned}
\mathcal{J}(\bm{x}_{j}) &= \dfrac{1}{2} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right)^{T} \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right) + \dfrac{\alpha_{\mathrm{reg}}}{2} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right)^{T} \bm{C}_{xx}^{-1} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right)\\
\mathcal{J}(\bm{x}_{j}) &= \mathcal{J}_{\mathrm{d}}(\bm{x}_{j}) + \mathcal{J}_{\mathrm{reg}}(\bm{x}_{j})
\end{aligned}
\end{equation}

We start by defining the prior ensemble of $N_{e}$ model realizations (ensemble members). This ensemble can be generated by various methods (MPS, conditionnal simulations etc.).

$$\bm{X} = \left(\bm{x}^{0}_{1}, \bm{x}^{0}_{2}, \dots, \bm{x}^{0}_{N_{e}} \right)$$

and we define the zero-mean (i.e., centered) anomaly matrix $\bm{A}$ as:

$$ \bm{A} = \bm{X}\left(\bm{I_{N_{e}}} - \dfrac{1}{N_{e}} \bm{11}^{T} \right) / \sqrt{N_{e}-1} $$

and we define the zero-mean (i.e., centered) anomaly matrix $\bm{A}$ as:

\begin{equation}
\begin{aligned}
\nabla \mathcal{J}(\bm{x}_{j}) & = \nabla g(\bm{x}_{j}) \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right) + \alpha_{\mathrm{reg}} \bm{C}_{xx}^{-1} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right)\\
& = \nabla g(\bm{x}_{j}) \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right) + \alpha_{\mathrm{reg}} \left(\bm{AA}^{T}\right)^{-1} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right)
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
\bm{H}_{j} & = \nabla g(\bm{x}_{j}) \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right) + \alpha_{\mathrm{reg}} \bm{C}_{xx}^{-1}\\
& = \nabla g(\bm{x}_{j}) \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right) + \alpha_{\mathrm{reg}} \left(\bm{AA}^{T}\right)^{-1}
\end{aligned}
\end{equation}

$$ \bm{H}_{j} x \approx \nabla^{2} \mathcal{J}_{\mathrm{d}}(\bm{x}_{j}) + \alpha_{\mathrm{reg}} \left(\bm{AA}^{T}\right)^{-1} \bm{x}_{j}$$

with the hessian $\nabla^{2} \mathcal{J}(x)$ applied to the vector $x$

So the adjoint direction can be obtained solving:

$$\bm{H}_{j}x_{j} = \nabla \mathcal{J}(\bm{x}_{j})$$

where $\bm{1} \in \mathbb{R}^{N_{e}}$ is defined as a vector with all elements equal to 1, $I_{N_e}$ is the $N$-dimensional identity matrix, and the projection $\bm{I_{N_{e}}} - \dfrac{1}{N_{e}} \bm{11}^{T}$ subtracts the mean from the ensemble. Thus, the ensemble covariance is:

$$ \bm{C}_{xx}^{-1} = \bm{AA}^{T}

Correspondingly, we can define an ensemble of perturbed measurements, $\bm{D} \in \mathbb{R}^{N_{\mathrm{obs}} \times N_e}$ , when given the real measurement vector, $d_{\mathrm{obs}} \in \mathbb{R}^{N_{obs}}$, as:

$$\bm{D} = \bm{d1}^{T} + \sqrt{N_{e} − 1}\bm{E},

where $\bm{E} \in \mathbb{R}^{N_{\mathrm{obs}} \times N_e}$ is the centered measurement-perturbation matrix whose columns are sampled from $\mathcal{N}(0, \bm{C}_{dd})$ and divided by $\sqrt{N_{e} − 1}$. Thus, we define the ensemble covariance matrix for the measurement perturbations as:

$$\bm{C}_{dd} = \bm{EE}^{T}.

## Cost Function in the Ensemble Subspace

It is easy to show that the EnRML solution is confined to the space spanned by the prior ensemble since the leftmost matrix in the gradient is the ensemble anomaly matrix. Thus, we will search for the solution in the ensemble subspace spanned by the prior ensemble by assuming that an updated ensemble realization at iteration $k$, $\bm{x}_{j}^{k+1}$, is equal to the prior realization, $\bm{x}_{j}^{k}$, plus a linear combination of the ensemble anomalies,

$$\bm{x}_{j}^{k+1} = \bm{x}_{j}^{k} + \bm{Aw}_{j}^{k+1}$$

In Raanes:

$$\bm{x}_{j}^{k+1} = \overline{\bm{x}}^{k} + \bm{Aw}_{j}^{k+1}$$


$$\mathcal{J}(\bm{x}_{j}) = \dfrac{1}{2} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right)^{T} \bm{C}_{xx}^{-1} \left(\bm{x}_{j} - \bm{x}_{j}^{f}\right) + \alpha_{\mathrm{reg}}\dfrac{1}{2} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right)^{T} \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}) - \bm{d}_{j}^{f}\right)$$

$$\mathcal{J}(\bm{w}_{j}) = \dfrac{1}{2} \bm{w}^{T} \bm{w}_{j} + \alpha_{\mathrm{reg}}\dfrac{1}{2} \left(g(\bm{x}_{j}^{k} + \bm{Aw}_{j}) - \bm{d}_{j}^{f}\right)^{T} \bm{C}_{dd}^{-1} \left(g(\bm{x}_{j}^{k} + \bm{Aw}_{j}) - \bm{d}_{j}^{f}\right)$$

$$\mathcal{J}(\bm{w}_{j}) = \dfrac{1}{2} \bm{w}^{T} \bm{w}_{j} + \dfrac{1}{2} \alpha_{\mathrm{reg}} \mathcal{J}_{\mathrm{d}}(\bm{w}_{j}) $$

Minimizing the cost functions in Equation (18) implies solving for the minima of the original cost functions in Equation (10), but restricted to the ensemble subspace and with $\bm{C}_{xx}$ in place of Cxx, as explained by Bocquet et al. [12]. The ensemble of cost functions in Equation (18), does not refer to the highdimensional state-covariance matrix, Cxx. The original problem, where the solution was searched for in the state space, is now reduced to a much simpler problem where we search for the solution in the ensemble space. Thus, we solve for the N vectors wj ∈ ℜN , one for each realization.

## Iterative Solution in the Ensemble Subspace

We will formulate a Gauss-Newton method for minimizing the cost function in Equation (18). The Jacobian (gradient) of the cost function 
$\nabla \mathcal{J}(\bm{w}_{j}) \in \mathbb{R}^{N_{e} \times 1}$ is:

With the adjoint state model, one knows how to compute 

$$ \nabla \mathcal{J}(\bm{x}_{j}) $$

$$\bm{X}_{i} = \bm{X} \left(\bm{I} + \bm{W}_{i} / \sqrt{N_{e} - 1} \right)$$

$$w_{j} = \bm{A}^{-1} \left(x_{j} - x_{f} \right)$$

$$\mathcal{J}(\bm{w}_{j}) = \dfrac{1}{2} \bm{w}^{T} \bm{w}_{j} + \alpha_{\mathrm{reg}} \mathcal{J}_{d}(\bm{w}_{j}) = \dfrac{1}{2} \bm{w}^{T} \bm{w}_{j} + \alpha_{\mathrm{reg}} \mathcal{J}_{\mathrm{d}, x}(\bm{x}_{j})$$

$$\dfrac{\mathcal{J(\bm{w}_{j})}}{\partial w_{j}} = \bm{w}_{j} + \dfrac{\partial \bm{x}_{j}}{\partial \bm{w}_{j}} \dfrac{\mathcal{J_{\mathrm{d}}(x_{j})}}{\partial \bm{x}_{j}} = \bm{w}_{j} + \left(\dfrac{\partial \bm{w}_{j}}{\partial \bm{x}_{j}}\right)^{-1} \dfrac{\mathcal{J_{2}(x_{j})}}{\partial \bm{x}_{j}} = \bm{w}_{j} + \bm{A} \dfrac{\mathcal{J_{2}(x_{j})}}{\partial \bm{x}_{j}} $$



$$\nabla \mathcal{J}(\bm{w}_{j}) = \bm{w}_{j} + \bm{A} \nabla \mathcal{J}(\bm{x}_{j})$$

And an approximate Hessian (gradient of the Jacobian) $\bm{H} \in \mathbb{R}^{N_{e} \times N_{e}}$ becomes:

$$\bm{H} (\bm{w}_j) \approx \bm{I} + (\bm{G}_{j}\bm{A})^{T}\bm{C}_{dd}^{-1} (\bm{G}_{j}\bm{A})$$ 

$(\bm{G}_{j}\bm{A})$ is estimated following evensen (2019).

Summary -> We obtain the gradient with the adjoint state and the Hessian from the ensemble.

The iterative Gauss-Newton scheme for minimization of the cost function in Equation (18) is:

$$ \bm{w}^{i+1}_{j} = \bm{w}_{j}^{i} − \gamma \left(\bm{H}(\bm{w}_{j}^{i})\right)^{− 1} \nabla \mathcal{J}(\bm{w}_{j}^{i}) = \bm{w}_{j}^{i} − \gamma \Bigg(\bm{I} + \left(\bm{G}_{j}^{i}\bm{A}\right)^{T}\bm{C}_{dd}^{-1} \left(\bm{G}_{j}^{i}\bm{A}\right) \Bigg)^{−1} \left(\bm{w}_{j}^{i} + \bm{A}^{T} \nabla \mathcal{J}(\bm{x}_{j})\right) $$

Now using the following two identities that can be derived from the Woodbury matrix lemma,

$$\left(\bm{I} + \bm{B}^{T}\bm{R}^{-1}\bm{B} \right)^{−1} = \bm{I} − \bm{B}^{T}\left(\bm{BB}^{T} + \bm{R}\right)^{−1}\bm{B},$$

we can write the Gauss-Newton iteration in Equation (22) as:

$$ \bm{w}^{i+1}_{j} = \bm{w}_{j}^{i} − \gamma \Bigg(\bm{I} − \left(\bm{G}_{j}^{i}\bm{A}\right)^{T}\left(\left(\bm{G}_{j}^{i}\bm{A}\right)\left(\bm{G}_{j}^{i}\bm{A}\right)^{T} + \bm{C}_{dd} \right)^{−1}\left(\bm{G}_{j}^{i}\bm{A}\right) \Bigg) \bm{A}^{T} \nabla \mathcal{J}(\bm{x}_{j})$$

When introducing the average sensitivity, $\overline{\bm{G}}_{i}$, for $\bm{G}_{j}^{i}$, and using Equations (29) and (38), we can write Equation (40) in vector form with $\bm{S}_{i}= \overline{\bm{G}}_{i}\bm{A}$:

In a matrix form, it reads

$$ \bm{W}^{i+1} = \bm{W}^{i} − \gamma \Bigg(\bm{I} − \bm{S}_{i}^{T}\left(\bm{S}_{i} \bm{S}_{i}^{T} + \bm{C}_{dd} \right)^{−1} \bm{S}_{i} \Bigg) \bm{A}^{T} \bm{P}$$

with $\bm{P}$ the ensemble of gradient, $\bm{p}_{j}$ being the $j^{th}$ gradient $\nabla \mathcal{J}(\bm{x}_{j})$.

### Exact inversion

With the use of the Woodbury corollary in Equation (25), we can write the iteration in Equation (42) as:

$$ \bm{W}^{i+1} = \bm{W}^{i} − \gamma \Bigg(\bm{I} − \left(\bm{S}_{i}^{T} \bm{C}_{dd}^{-1} \bm{S}_{i} + \bm{I}_{N_e} \right)^{−1} \bm{S}_{i}^{T} \bm{C}_{dd}^{-1} \bm{S}_{i} \Bigg) \bm{A}^{T} \bm{P}$$

where the inversion is computed in the ensemble space (see also Equation 22). In the case where either C−1 dd or the square root C− 1 2 dd is known we can compute the product SiTC−1 dd Si at a cost of 2m2N operations, and the inverse is computed at a cost of O(N3) operations. For the case of a diagonal error covariance matrix Cdd = Im, Equation (50) reduces to:

$$ \bm{W}^{i+1} = \bm{W}^{i} − \gamma \Bigg(\bm{I} − \left(\bm{S}_{i}^{T} \bm{S}_{i} + \bm{I}_{N_e} \right)^{−1} \bm{S}_{i}^{T} \bm{S}_{i} \Bigg) \bm{A}^{T} \bm{P}$$
