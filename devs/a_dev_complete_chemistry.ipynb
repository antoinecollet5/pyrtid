# Governing equations

In this work, we are interested in the model describing density driven transient saturated flow in heterogeneous porous media (case of captive water table) associated to transport and chemical reactions of first order kinetics.

## Continuous form

 In continuous form, the governing equations are written as follows (\cref{eq:fwd_system_continuous}),

\begin{equation}
    \left\{
        \begin{array}{ll}
\mathcal{S}_{s} \dfrac{\partial p}{\partial t} = \nabla\cdot\left(K \cdot \overrightarrow{\nabla} \left(p + \rho g\right) \right) - q\rho, \quad \text{in} \; T, \Omega_{\mathrm{int}}
        \\
		h = \dfrac{p}{\rho g} + z, \quad  \text{in} \; T, \Omega
		\\
		\dfrac{\partial h}{\partial t} = 0, \quad \text{in} \; T, \partial \Omega
		\\ 
		p^{0} = \mathrm{cst}, \quad \text{in} \; \Omega
        \end{array}
\right.
% \label{eq:fwd_system_continuous}
\end{equation}

# Discrete form

Using Voronoi-type FVM for spatial discretization \citep{vanderleeModuleorientedModelingReactive2003}, all state variables are defined at the mesh centers.
Pressures and  concentrationsthe in mesh $i$ are expressed as functions of the state variables in neighboring meshes $j$ (\cref{fig:fv_voronoi_scheme}). The discrete equation form is independent of domain dimensionality and is \cref{eq:fwd_eq_fv_discrete}:


\begin{equation}
\left\{ \begin{array}{lll}
    (1) & \begin{array}{l} \forall\ n \in [0, N_{t}-1] \\ \forall\ i \in \Omega_{\mathrm{int}} \end{array}, & \begin{array}{l}\begin{aligned} V_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = & \sum_{neigh \;j} \left\lVert \Gamma_{ij}  \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + \rho_{ij}^{n} g \cdot \overrightarrow{1_z} \cdot \overrightarrow{n_{ij}} \Bigg)
          \\
          & - \alpha_{\mathrm{f}} Q_{i}^{n}\rho_{i}^{n} - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1}\rho_{i}^{n} \end{aligned}\end{array}
    \\\\
    (2) & \begin{array}{l} \forall\ n \in [0, N_{t}] \\ \forall\ i \in \Omega \end{array}, & h_{i}^{n} = \dfrac{p_{i}^{n}}{\rho_{i}^{n}g} + z_{i}
    \\\\
    (3) & \begin{array}{l} \forall\ n \in [0, N_{t}-1] \\ \forall\ i \in \Omega \end{array}, & \rho_{i}^{n+1} = \rho_{0} \left(1 + \sum_{sp} c_{sp, i}^{n} M_{sp} \right)
    \\\\
    (5) & \begin{array}{l} \forall\ n \in [0, N_{t}-1] \\ \forall\ i \in \Omega_{\mathrm{int}} \end{array}, & \begin{array}{l}\begin{aligned} V_{i} \omega_{e, i} \dfrac{c_{i}^{n+1} - c_{i}^{n}}{\Delta t^{n}} & = \displaystyle \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert D_{e, ij}\Bigg(\alpha_{\mathrm{d}}\dfrac{c_{j}^{n+1} - c_{i}^{n+1}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{d}})\dfrac{c_{j}^{n} - c_{i}^{n}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\Bigg) \\
      & - V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{n+1} - \overline{c}_{i}^{n}}{\Delta t^{n}} \end{aligned}\end{array}
    \\\\
    (6) & \begin{array}{l} \forall\ n \in [0, N_{t}-1] \\ \forall\ i \in \Omega_{\mathrm{int}} \end{array}, & V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{n+1} - \overline{c}_{i}^{n}}{\Delta t^{n}} = - V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{n} \left( 1 - \dfrac{c_{i}^{n+1}}{K_{s}}\right)

    \\\\
    (3) & \begin{array}{l} \forall\ n \in [0, N_{t}] \\ \forall\ i \in  \partial \Omega \end{array}, &  p_{i}^{n+1} = p_{i}^{n}
    \\\\
  	(4) & \forall\ i \in \Omega , &\left\{\begin{array}{ll}
    \text{if stationary}, &\displaystyle\sum\limits_{\mathrm{neigh} \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{\mathbf{P}_{i}\mathbf{P}_{j}} \right\rVert} - Q_{i}^{0} = 0 \\ \text{else}, & p_{i}^{0} = \text{cst}
    \end{array}\right.
    \\\\
    (7) & \begin{array}{l} \forall\ n \in [0, N_{t}] \\ \forall\ i \in  \partial \Omega \end{array}, &  c_{i}^{n+1} = c_{i}^{n}
    \\\\
    (8) & \begin{array}{l} \forall\ n \in [0, N_{t}] \\ \forall\ i \in  \partial \Omega \end{array}, &  \overline{c}_{i}^{n+1} = \overline{c}_{i}^{n}
    \\\\
    (9) & \forall\ i \in \Omega, &  c_{i}^{0} = \mathrm{cst}, \quad \overline{c}_{i}^{0} = \mathrm{cst}, \quad \rho_{i}^{0} = \mathrm{cst}
    \\\\
\end{array}\right.
%\label{eq:fwd_eq_fv_discrete}
\end{equation}
