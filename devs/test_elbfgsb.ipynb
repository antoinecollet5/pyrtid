```python
# python lbfgsb works with numpy arrays
import numpy as np
from pyrtid.inverse import minimize_ensemble_lbfgsb
from scipy.optimize import minimize
```

- First example: min x^{\mathrm{T}}x such that x>=1. Optimal solution hits the boundary.

```python
def quad(x):
    # return x.dot(x)
    return 10 * x[0] ** 2 + x[1] ** 2


def grad_quad(x):
    return np.array([10 * x[0], x[1]])


def fun_and_grad(x):
    return quad(x), grad_quad(x)


ftol = 1e-5
gtol = 1e-5
l = np.array([1.0, 1.0])
u = np.array([np.inf, np.inf])
bounds = np.array((l, u)).T
x0 = np.array([5, 5])

# Perform a normal sampling
x0 = np.random.random(size=[2, 20]) * 4.0 + 3.0
```

```python
x0
```

```python
bounds[0]
```

```python
x0[:, 0]
```

### New implementation

```python
print("====================== Quadratic example ======================")
opt_quad = minimize_ensemble_lbfgsb(
    x0=x0,
    fun_and_jac=fun_and_grad,
    bounds=bounds,
    ftol=ftol,
    gtol=gtol,
    iprint=101,
    maxcor=40,
    max_iter=5,
    is_initialize_bfgs_matrices_from_ensemble=True,
)
print("")
print("")
xOpt = np.array([1, 1])
theoOpt = {"x": xOpt, "f": quad(xOpt), "df": grad_quad(xOpt)}
```

```python
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
print(opt_quad)
```

```python
opt_quad.x
```

## Second example : min Rosenbrock function

```python
def rosenbrock(x):
    return 100 * (x[1] - x[0] ** 2) ** 2 + (1 - x[0]) ** 2


def grad_rosenbrock(x):
    g = np.empty(x.size)
    g[0] = 400 * x[0] * (x[0] ** 2 - x[1]) + 2 * (x[0] - 1)
    g[1] = 200 * (-x[0] ** 2 + x[1])
    return g


def fun_and_grad_rosenbrock(x):
    return rosenbrock(x), grad_rosenbrock(x)


l = np.array([-2, -2])
u = np.array([2, 2])
bounds = np.array((l, u)).T
# x0 = np.array([0.12, 0.12])
# x0 = np.array([-1, -1])

# x0 = (np.random.random(size=[2, 10]) - 0.5) * 4
x0 = np.random.random(size=[2, 2]) * 0.0 + -1.0
```

```python
x0
```

### New implementation

```python
np.finfo(1.0).resolution * 1e7
```

```python
opt_rosenbrock = minimize_ensemble_lbfgsb(
    x0=x0,
    fun_and_jac=fun_and_grad_rosenbrock,
    bounds=bounds,
    ftol=ftol,
    gtol=gtol,
    iprint=101,
    ftol_linesearch=1e-3,
    gtol_linesearch=0.9,
    xtol_linesearch=0.1,
    is_initialize_bfgs_matrices_from_ensemble=True,
)
print("")
print("")
theoOpt = {"x": np.array([1, 1]), "f": 0, "df": grad_rosenbrock(np.array([1, 1]))}
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
```

```python
print(opt_rosenbrock)
```

## Third example : min Beale function

```python
def beale(x):
    return (
        (1.5 - x[0] + x[0] * x[1]) ** 2
        + (2.25 - x[0] + x[0] * x[1] ** 2) ** 2
        + (2.625 - x[0] + x[0] * x[1] ** 3) ** 2
    )


def grad_beale(x):
    y1 = x[1]
    y2 = y1 * y1
    y3 = y2 * y1
    f1 = 1.5 - x[0] + x[0] * y1
    f2 = 2.25 - x[0] + x[0] * y2
    f3 = 2.625 - x[0] + x[0] * y3

    return np.array(
        [
            2 * (y1 - 1) * f1 + 2 * (y2 - 1) * f2 + 2 * (y3 - 1) * f3,
            2 * x[0] * f1 + 4 * x[0] * y1 * f2 + 6 * x[0] * y2 * f3,
        ]
    )


def beale_fun_and_jac(x):
    return beale(x), grad_beale(x)


ftol = 1e-14
gtol = 1e-10
l = -4.5 * np.ones(2)
u = -l
bounds = np.array((l, u)).T
# x0 = np.array([2.5, -1.3])

# Equivalent to the vector version.
# x0 = np.array([[2.5],[-1.3]])

# This is more complex
# x0 = (np.random.random(size=[2, 10]) - 0.5) * 4
ne = 30
x0 = np.zeros((2, ne))
x0[0] = np.random.random(ne) * 2 + 2.0
x0[1] = np.random.random(ne) * 1 - 1.0
```

```python
x0
```

### New implementation

```python
opt_beale = minimize_ensemble_lbfgsb(
    x0=x0,
    fun_and_jac=beale_fun_and_jac,
    bounds=bounds,
    ftol=ftol,
    gtol=gtol,
    maxcor=20,
    is_initialize_bfgs_matrices_from_ensemble=True,
    max_iter=9,
)
print("")
print("")
theoOpt = {
    "x": np.array([3, 0.5]),
    "f": beale(np.array([3, 0.5])),
    "df": grad_beale(np.array([3, 0.5])),
}
print("Theoretical optimal value: ")
print(theoOpt)
print("Optimal value found: ")
```
