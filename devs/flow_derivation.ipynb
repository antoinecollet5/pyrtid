# Dérivation de l'équation d'écoulement saturé

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
      \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} - Q_{i}^{n} = 0 
    \end{array}
  \right.
$$


Et en regime transitoire avec le schéma de crank-nicolson

$$
  \forall n \in [0, N], \left\{
    \begin{array}{l}
      \mathcal{A}_{i} {S_{S}}_{i} \dfrac{h_{i}^{n+1} - h_{i}^{n}}{\Delta t^{n}} = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \Bigg(K_{ij}\alpha_{\mathrm{f}}\dfrac{h_{j}^{n+1} - h_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{h_{j}^{n} - h_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) - \alpha_{\mathrm{f}} Q_{i}^{n} - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1} 
      \\
      p_{i}^{n} = (h_{i}^{n} - z_{i})\rho_{i}^{n}g
    \end{array}
  \right.
$$


$$ \mathcal{L} = \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\sigma_{p_{i}^{n, \mathrm{obs}}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\sigma_{h_{i}^{n, \mathrm{obs}}}}\right)^{2}- \langle \lambda_{p_{i}^{n+1}}, g_{p} \rangle$$

Explicitely it reads:

\begin{equation}
\begin{aligned}
\mathcal{L} = &\dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\sigma_{p_{i}^{n, \mathrm{obs}}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\sigma_{h_{i}^{n, \mathrm{obs}}}}\right)^{2}
\\
&- \sum_{n=0}^{N} \lambda_{h_{i}}^{n+1}\Bigg[\mathcal{A}_{i} {S_{S}}_{i} \dfrac{h_{i}^{n+1} - h_{i}^{n}}{\Delta t^{n}} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \Bigg(K_{ij}\alpha_{\mathrm{f}}\dfrac{h_{j}^{n+1} - h_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{h_{j}^{n} - h_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) + \alpha_{\mathrm{f}} Q_{i}^{n} + (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1} \Bigg]
\\
&- \sum_{n=0}^{N+1} \lambda_{p_{i}}^{n} \Bigg[p_{i}^{n} + (h_{i}^{n} - z_{i})\rho_{i}^{n}g\Bigg]  
\end{aligned}
\end{equation}

Note: il est plus aisé d'obtenir une variable supplémentaire que de faire de multiples changements entre h et p. Qi = flowrates in m3/s

## Dérivation par $h_{i}^{n}$

$$
\dfrac{\partial \mathcal{L}}{\partial \overline{c}_{i}^{n}} =
\begin{cases}
    - \dfrac{h_{i}^{0, \mathrm{obs}} - h_{i}^{0, \mathrm{calc}}}{\left(\sigma_{h_{i}}^{0, \mathrm{obs}}\right)^{2}} + \dfrac{\mathcal{A}_{i} {S_{S}}_{i}}{\Delta t^{0}} \lambda_{h_{i}}^{1} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}(1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{h_{j}}^{1} - \lambda_{h_{i}}^{1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \lambda_{p_{i}}^{0} \rho_{i}^{0}g, & \text{if}\ n = 0 
    \\\\
    - \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\left(\sigma_{h_{i}}^{n, \mathrm{obs}}\right)^{2}} + \mathcal{A}_{i} {S_{S}}_{i} \left(\dfrac{\lambda_{p_{i}}^{n+1}}{\Delta t^{n}} - \dfrac{\lambda_{p_{i}}^{n}}{\Delta t^{n-1}}\right) + \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \Bigg(K_{ij}\alpha_{\mathrm{f}}\dfrac{\lambda_{h_{j}}^{n} - \lambda_{h_{i}}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{h_{j}}^{n+1} - \lambda_{h_{i}}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) , & \forall n \in [1, N] 
    \\\\
    - \dfrac{h_{i}^{N+1, \mathrm{obs}} - h_{i}^{N+1, \mathrm{calc}}}{\left(\sigma_{h_{i}}^{N+1, \mathrm{obs}}\right)^{2}} - \dfrac{\mathcal{A}_{i} {S_S}_{i}}{\Delta t^{N}}  \lambda_{h_{i}}^{N+1} + \lambda_{p_{i}}^{N+1} \rho_{i}^{N+1}g, & \text{if}\ n = N+1
\end{cases}
$$

Since we want to impose $\dfrac{\partial \mathcal{L}}{\partial \overline{c}_{i}^{n}} = 0$, it gives the initial condition for the adjoint state:

$$ \lambda_{\overline{c}_{i}}^{N+1} =  - \dfrac{\mathcal{A}_{i} \omega_{i}}{\Delta t^{N}} \lambda_{c_{i}}^{N+1} - \dfrac{\overline{c}_{i}^{N+1, \mathrm{obs}} - \overline{c}_{i}^{N+1, \mathrm{calc}}}{\left(\sigma_{\overline{c}_{i}}^{N+1, \mathrm{obs}}\right)^{2}}$$

And the temporal evolution, $\mathcal{A}_{i} \omega_{i} \left(\dfrac{\lambda_{c_{i}}^{n}}{\Delta t^{n-1}} - \dfrac{\lambda_{c_{i}}^{n+1}}{\Delta t^{n}}\right)$ being an "adjoint transport" error source term:

$$\lambda_{\overline{c}_{i}}^{n} = \lambda_{\overline{c}_{i}}^{n+1} \left(1 + \Delta t^{n} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{n+1}}{K_{s}}\right) \right) - \mathcal{A}_{i} \omega_{i} \left(\dfrac{\lambda_{c_{i}}^{n}}{\Delta t^{n-1}} - \dfrac{\lambda_{c_{i}}^{n+1}}{\Delta t^{n}}\right) - \dfrac{\overline{c}_{i}^{n, \mathrm{obs}} - \overline{c}_{i}^{n, \mathrm{calc}}}{\left(\sigma_{\overline{c}_{i}}^{n, \mathrm{obs}}\right)^{2}}, \forall n \in [1, N] $$
