# Dérivation de l'équation d'écoulement saturé

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
      \displaystyle\sum\limits_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} - Q_{i}^{0} = 0 
    \end{array}
  \right.
$$


Et en regime transitoire avec le schéma de crank-nicolson

$$
  \forall n \in [0, N], \left\{
    \begin{array}{l}
      \mathcal{A}_{i} {S_{S}}_{i} \dfrac{h_{i}^{n+1} - h_{i}^{n}}{\Delta t^{n}} = \displaystyle\sum\limits_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{h_{j}^{n+1} - h_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{h_{j}^{n} - h_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) - \alpha_{\mathrm{f}} Q_{i}^{n} - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1} 
      \\
      p_{i}^{n} = (h_{i}^{n} - z_{i})\rho_{i}^{n}g
    \end{array}
  \right.
$$


The total head or hydraulic head = pressure head [L] + elevation + Kinetic term ($\mu^{2} / 2g $) that is neglected here.

$$ \mathcal{L} = \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{\mathrm{obs}, i}^{n} - p_{\mathrm{calc}, i}^{n}}{\sigma_{p_{\mathrm{obs}, i}^{n}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{\mathrm{obs}, i}^{n} - h_{\mathrm{calc}, i}^{n}}{\sigma_{h_{\mathrm{obs}, i}^{n}}}\right)^{2} + \dfrac{1}{2} \sum_{i} \left(\dfrac{K_{\mathrm{obs}, i} - K_{\mathrm{calc}, i}}{\sigma_{K_{\mathrm{obs}, i}}}\right)^{2} - \langle \lambda_{h_{i}^{n+1}}, g_{h} \rangle - \langle \lambda_{p_{i}^{n+1}}, g_{p} \rangle$$

Explicitely it reads:

\begin{equation}
\begin{aligned}
\mathcal{L} = &\dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{\mathrm{obs}, i}^{n} - p_{\mathrm{calc}, i}^{n}}{\sigma_{p_{\mathrm{obs}, i}^{n}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{\mathrm{obs}, i}^{n} - h_{\mathrm{calc}, i}^{n}}{\sigma_{h_{\mathrm{obs}, i}^{n}}}\right)^{2} + \dfrac{1}{2} \sum_{i} \left(\dfrac{K_{\mathrm{obs}, i} - K_{\mathrm{calc}, i}}{\sigma_{K_{\mathrm{obs}, i}}}\right)^{2} 
\\
&- \sum_{n=0}^{N} \lambda_{h_{i}}^{n+1}\Bigg[\mathcal{A}_{i} {S_{S}}_{i} \dfrac{h_{i}^{n+1} - h_{i}^{n}}{\Delta t^{n}} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{h_{j}^{n+1} - h_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{h_{j}^{n} - h_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) + \alpha_{\mathrm{f}} Q_{i}^{n} + (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1} \Bigg]
\\
&- \sum_{n=0}^{N+1} \lambda_{p_{i}}^{n} \Bigg[p_{i}^{n} + (h_{i}^{n} - z_{i})\rho_{i}^{n}g\Bigg]  
\end{aligned}
\end{equation}

Note: il est plus aisé d'obtenir une variable supplémentaire que de faire de multiples changements entre h et p. Qi = flowrates in m3/s

## Dérivation par $h_{i}^{n}$

\begin{equation}
\dfrac{\partial \mathcal{L}}{\partial h_{i}^{n}} =
\begin{cases}
    - \dfrac{h_{\mathrm{obs}, i}^{0} - h_{\mathrm{calc}, i}^{0}}{\left(\sigma_{h_{\mathrm{obs}, i}}^{0}\right)^{2}} + \dfrac{\mathcal{A}_{i} \mathcal{S}_{s_i}}{\Delta t^{0}} \lambda_{h_{i}}^{1} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}(1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{h_{j}}^{1} - \lambda_{h_{i}}^{1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \lambda_{p_{i}}^{0} \rho_{i}^{0}g, & \text{if}\ n = 0 
    \\\\
    \!\begin{aligned}%[b]
    &- \dfrac{h_{\mathrm{obs}, i}^{n} - h_{\mathrm{calc}, i}^{n}}{\left(\sigma_{h_{\mathrm{obs}, i}}^{n}\right)^{2}} + \mathcal{A}_{i} \mathcal{S}_{s_i} \left(\dfrac{\lambda_{h_{i}}^{n+1}}{\Delta t^{n}} - \dfrac{\lambda_{h_{i}}^{n}}{\Delta t^{n-1}}\right) \\
    & \quad \quad + \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{\lambda_{h_{j}}^{n} - \lambda_{h_{i}}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{h_{j}}^{n+1} - \lambda_{h_{i}}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg)  
    + \lambda_{p_{i}}^{n} \rho_{i}^{n}g
    \end{aligned}
    , & \forall n \in [1, N] 
    \\\\
    - \dfrac{h_{\mathrm{obs}, i}^{N_{ts}+1} - h_{\mathrm{calc}, i}^{N_{ts}+1}}{\left(\sigma_{h_{\mathrm{obs}, i}}^{N_{ts}+1}\right)^{2}} - \dfrac{\mathcal{A}_{i} {S_S}_{i}}{\Delta t^{N_{ts}}}  \lambda_{h_{i}}^{N_{ts}+1} + \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\alpha_{\mathrm{f}}\dfrac{\lambda_{h_{j}}^{N_{ts}+1} - \lambda_{h_{i}}^{N_{ts}+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \lambda_{p_{i}}^{N_{ts}+1} \rho_{i}^{N_{ts}+1}g, & \text{if}\ n = N+1
\end{cases}
\end{equation}

Since we want to impose $\dfrac{\partial \mathcal{L}}{\partial h_{i}^{n}} = 0$, it gives the initial condition for the adjoint state:

\begin{equation}
\dfrac{\mathcal{A}_{i} \mathcal{S}_{s_i}}{\Delta t^{N_{ts}}}  \lambda_{h_{i}}^{N_{ts}+1} = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\alpha_{\mathrm{f}}\dfrac{\lambda_{h_{j}}^{N_{ts}+1} - \lambda_{h_{i}}^{N_{ts}+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \lambda_{p_{i}}^{N_{ts}+1} \rho_{i}^{N_{ts}+1}g - \dfrac{h_{\mathrm{obs}, i}^{N_{ts}+1} - h_{\mathrm{calc}, i}^{N_{ts}+1}}{\left(\sigma_{h_{\mathrm{obs}, i}}^{N_{ts}+1}\right)^{2}}
\end{equation}

And the temporal evolution:

\begin{equation}
\mathcal{A}_{i} \mathcal{S}_{s_i} \left(\dfrac{\lambda_{p_{i}}^{n}}{\Delta t^{n-1}} -\dfrac{\lambda_{p_{i}}^{n+1}}{\Delta t^{n}}\right) = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{\lambda_{h_{j}}^{n} - \lambda_{h_{i}}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{h_{j}}^{n+1} - \lambda_{h_{i}}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) + \lambda_{p_{i}}^{n} \rho_{i}^{n}g - \dfrac{h_{\mathrm{obs}, i}^{n}  - h_{\mathrm{calc}, i}^{n}}{\left(\sigma_{h_{\mathrm{obs}, i}}^{n}\right)^{2}}, \forall n \in [1, N] 
\end{equation}

## Dérivation par $p_{i}^{n}$

$$
\dfrac{\partial \mathcal{L}}{\partial p_{i}^{n}} = - \dfrac{p_{\mathrm{obs}, i}^{n} - p_{\mathrm{calc}, i}^{n}}{\left(\sigma_{p_{\mathrm{obs}, i}}^{n}\right)^{2}} - \lambda_{p_{i}}^{n},\forall n \in [0, N+1] 
$$

Since we want to impose $\dfrac{\partial \mathcal{L}}{\partial p_{i}^{n}} = 0$

$$ \implies \lambda_{p_{i}}^{n} = - \dfrac{p_{\mathrm{obs}, i}^{n} - p_{\mathrm{calc},i}^{n}}{\left(\sigma_{p_{\mathrm{obs},i}}^{n}\right)^{2}},\forall n \in [0, N+1] $$

## Derivation by $K_{i}$

$$\dfrac{\partial \mathcal{L}}{\partial K_{i}} = \dfrac{\partial K_{ij}}{\partial K_{i}}\sum_{n=0}^{N} \sum_{neigh \;j}  \dfrac{\left\lVert \Gamma_{ij}  \right\rVert}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} \Bigg(\alpha_{\mathrm{f}}\left(h_{j}^{n+1} - h_{i}^{n+1}\right) + (1 - \alpha_{\mathrm{f}})\left(h_{j}^{n} - h_{i}^{n}\right)\Bigg)\left(\lambda_{h_{j}}^{n+1} - \lambda_{h_{i}}^{n+1} \right) - \dfrac{K_{i}^{\mathrm{obs}} - K_{i}^{\mathrm{calc}}}{\left(\sigma_{K_{i}}^{\mathrm{obs}}\right)^{2}}$$
