# Dérivation de l'équation d'écoulement densitaire

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
      \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} - Q_{i}^{n} = 0 
    \end{array}
  \right.
$$


Et en regime transitoire avec le schéma de crank-nicolson

$$
  \forall n \in [0, N-1], \left\{
    \begin{array}{l}
      \mathcal{A}_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = \sum_{neigh \;j} \left\lVert \Gamma_{ij}  \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} \Bigg(K_{ij}\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) - \alpha_{\mathrm{f}} Q_{i}^{n}\rho_{i}^{n} - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1}\rho_{i}^{n}
      \\
      h_{i}^{n} - = \dfrac{p_{i}^{n}}{\rho_{i}^{n}} + z_{i}
    \end{array}
  \right.
$$


Note: Nous utilisons ici $\rho_{i}^{n}$ et non $\rho_{i}^{n+1}$ car l'écoulement est calculé avant le transport et donc la densité n'est pas encore connue. 

$$ \mathcal{L} = \dfrac{1}{2} \sum_{n=0}^{N} \sum_{i} \left( \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\sigma_{p_{i}^{n, \mathrm{obs}}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N} \sum_{i} \left( \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\sigma_{h_{i}^{n, \mathrm{obs}}}}\right)^{2}- \langle \lambda_{p_{i}^{n+1}}, g_{p} \rangle$$
