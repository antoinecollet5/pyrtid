# Dérivation de l'équation d'écoulement densitaire

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
      \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} - Q_{i}^{n} = 0 
    \end{array}
  \right.
$$


$$
      V_{i} {S_{S}}_{i} \left(p_{i}^{n+1} - p_{i}^{n}\right){\Delta t^{n}} = 0
$$

### 1) Initialization of the system (potential saturated flow)

$$
  n=0 \implies \left\{
    \begin{array}{llll}
      1.1)& F_{c}^{0} &:& c^{0}_{i, sp} = \mathrm{cst} \\
      1.2)& F_{\overline{c}}^{0} &:& \overline{c}^{0}_{i, sp} = \mathrm{cst} \\
      1.3)& F_{p}^{0} &:& p_{i}^{0} = \mathrm{cst} \\ % or stationray
      2) & F_{\rho}^{0} &: & \rho_{i}^{0} = \rho_{0} \left(1 + \displaystyle\sum_{sp} c_{i, sp}^{0} M_{sp} \right) \\
      3) & F_{U}^{0} &: & \overrightarrow{U}_{ij}^{n} = - \dfrac{K_{ij}}{\rho_{0}g} \dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}}\right\rVert} \cdot  \overrightarrow{n_{ij}} + \rho_{ij}^{n-1} \cdot \overrightarrow{z}\\
      4) & F_{h}^{0} & :& h_{i}^{0} = \dfrac{p_{i}^{0}}{\rho_{i}^{0}g} + z_{i} 
     
    \end{array}
  \right.
$$

### 2) Reactive transport iterations

$$
  \forall n \in [0, N_{ts}-1], \left\{
    \begin{array}{ll}
    1) F_{p}: &  V_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = \displaystyle \sum_{neigh \;j} \left\lVert \Gamma_{ij}  \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) - \alpha_{\mathrm{f}} Q_{i}^{n+1}\rho_{i}^{n}g - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n}\rho_{i}^{n}g
      \\
      2) F_{h}: & h_{i}^{n+1} = \dfrac{p_{i}^{n+1}}{\rho_{i}^{n}g} + z_{i}
      \\
      3) F_{U} : & \overrightarrow{U}_{ij}^{n+1} = - \dfrac{K_{ij}}{\rho_{0}g} \dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}}\right\rVert} \cdot \overrightarrow{n_{ij}} + \rho_{ij}^{n} \cdot \overrightarrow{z}
      \\
      4) F_{c}: & \begin{aligned}
 V_{i} \omega_{e,i} \dfrac{c_{i}^{n+1} - c_{i}^{n}}{\Delta t^{n}} & = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert D_{e, ij}\Bigg(\alpha_{\mathrm{d}}\dfrac{c_{j}^{n+1} - c_{i}^{n+1}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{d}})\dfrac{c_{j}^{n} - c_{i}^{n}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\Bigg)
\\
& \quad - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \Bigg(\alpha_{\mathrm{a}} c_{ij}^{n+1} \overrightarrow{U_{ij}^{n+1}} + (1 - \alpha_{\mathrm{a}}) c_{ij}^{n} \overrightarrow{U_{ij}^{n}} \Bigg) \cdot \overrightarrow{n_{ij}}
\\
& \quad + \sum_{neigh \;j}  \left\lVert \Gamma_{ij} \right\rVert \left( \alpha_{\mathrm{a}} \left( \overrightarrow{U_{ij}^{n+1}} \cdot  \overrightarrow{n_{ij}} - Q_{i}^{n+1} \right) c_{i}^{n+1}  + (1 - \alpha_{\mathrm{a}}) \left( \overrightarrow{U_{ij}^{n}} \cdot  \overrightarrow{n_{ij}} - Q_{i}^{n} \right) c_{i}^{n}  \right)
\\
& \quad + \alpha_{\mathrm{a}} Q_{i, \mathrm{out}}^{n+1} c_{i}^{n+1} + (1 - \alpha_{\mathrm{a}}) Q_{i, \mathrm{out}}^{n} c_{i}^{n} + Q_{i,\mathrm{in}}^{n} c_{i, \mathrm{in}}^{n}
\\
& \quad - V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{n+1} - \overline{c}_{i}^{n}}{\Delta t^{n}}
\end{aligned}
\\
5) F_{\overline{c}}: & V_{i} \omega_{e,i} \dfrac{\overline{c}_{i}^{n+1} - \overline{c}_{i}^{n}}{\Delta t^{n+1}} = V_{i} \omega_{e,i} k_{v} A_{s} \overline{c}_{i}^{n} \left( 1 - \dfrac{c_{i}^{n}}{K_{s}}\right)
\\
6) F_{\rho}: & \rho_{i}^{n+1} = \rho_{0} \left(1 + \displaystyle\sum_{sp} c_{i, sp}^{n+1} M_{sp} \right)
\end{array}
\right.
$$
