# Dérivation de l'équation d'écoulement densitaire

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
      \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} - Q_{i}^{n} = 0 
    \end{array}
  \right.
$$


Et en regime transitoire avec le schéma de crank-nicolson

$$
  \forall n \in [0, N-1], \left\{
    \begin{array}{l}
      \mathcal{A}_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = \sum_{neigh \;j} \left\lVert \Gamma_{ij}  \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) - \alpha_{\mathrm{f}} Q_{i}^{n}\rho_{i}^{n} - (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1}\rho_{i}^{n}
      \\
      h_{i}^{n} = \dfrac{p_{i}^{n}}{\rho_{i}^{n}g} + z_{i}
    \end{array}
  \right.
$$


Note: Nous utilisons ici $\rho_{i}^{n}$ et non $\rho_{i}^{n+1}$ car l'écoulement est calculé avant le transport et donc la densité n'est pas encore connue. 

$$ \mathcal{L} = \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\sigma_{p_{i}^{n, \mathrm{obs}}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\sigma_{h_{i}^{n, \mathrm{obs}}}}\right)^{2} - \langle \lambda_{h_{i}^{n+1}}, g_{h} \rangle - \langle \lambda_{p_{i}^{n+1}}, g_{p} \rangle$$

\begin{equation}
\begin{aligned}
\mathcal{L} = &\dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\sigma_{p_{i}^{n, \mathrm{obs}}}}\right)^{2} + \dfrac{1}{2} \sum_{n=0}^{N+1} \sum_{i} \left( \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\sigma_{h_{i}^{n, \mathrm{obs}}}}\right)^{2}
\\
&- \sum_{n=0}^{N} \lambda_{p_{i}}^{n+1}\Bigg[\mathcal{A}_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} - \sum_{neigh \;j} \left\lVert \Gamma_{ij}  \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) + \alpha_{\mathrm{f}} Q_{i}^{n}\rho_{i}^{n} + (1 - \alpha_{\mathrm{f}}) Q_{i}^{n-1}\rho_{i}^{n} \Bigg]
\\
&- \sum_{n=0}^{N+1} \lambda_{h_{i}}^{n} \Bigg[h_{i}^{n} - \dfrac{p_{i}^{n}}{\rho_{i}^{n}g} - z_{i}\Bigg]  
\end{aligned}
\end{equation}

## Derivation by $p_{i}^{n}$

$$
\dfrac{\partial \mathcal{L}}{\partial h_{i}^{n}} =
\begin{cases}
    - \dfrac{p_{i}^{0, \mathrm{obs}} - p_{i}^{0, \mathrm{calc}}}{\left(\sigma_{p_{i}}^{0, \mathrm{obs}}\right)^{2}} + \dfrac{\mathcal{A}_{i} {S_{S}}_{i}}{\Delta t^{0}} \lambda_{p_{i}}^{1} - \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \dfrac{\rho_{ij}^{0}}{\rho_{0}} K_{ij}(1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{p_{j}}^{1} - \lambda_{p_{i}}^{1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \dfrac{\lambda_{h_{i}}^{0}}{\rho_{i}^{0}g}, & \text{if}\ n = 0 
    \\\\
    - \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\left(\sigma_{p_{i}}^{n, \mathrm{obs}}\right)^{2}} + \mathcal{A}_{i} {S_{S}}_{i} \left(\dfrac{\lambda_{p_{i}}^{n+1}}{\Delta t^{n}} - \dfrac{\lambda_{p_{i}}^{n}}{\Delta t^{n-1}}\right) + \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{\lambda_{p_{j}}^{n} - \lambda_{p_{i}}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{p_{j}}^{n+1} - \lambda_{p_{i}}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) + \dfrac{\lambda_{h_{i}}^{n}}{\rho_{i}^{n}g}, & \forall n \in [1, N] 
    \\\\
    - \dfrac{p_{i}^{N+1, \mathrm{obs}} - p_{i}^{N+1, \mathrm{calc}}}{\left(\sigma_{p_{i}}^{N+1, \mathrm{obs}}\right)^{2}} - \dfrac{\mathcal{A}_{i} {S_S}_{i}}{\Delta t^{N}}  \lambda_{p_{i}}^{N+1} + \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \alpha_{\mathrm{f}}\dfrac{\lambda_{p_{j}}^{N+1} - \lambda_{p_{i}}^{N+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \dfrac{\lambda_{h_{i}}^{N+1}}{\rho_{i}^{N+1}g}, & \text{if}\ n = N+1
\end{cases}
$$

Since we want to impose $\dfrac{\partial \mathcal{L}}{\partial h_{i}^{n}} = 0$, it gives the initial condition for the adjoint state:

$$\dfrac{\mathcal{A}_{i} {S_S}_{i}}{\Delta t^{N}}  \lambda_{p_{i}}^{N+1} = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \alpha_{\mathrm{f}}\dfrac{\lambda_{p_{j}}^{N+1} - \lambda_{p_{i}}^{N+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + \dfrac{\lambda_{h_{i}}^{N+1}}{\rho_{i}^{N+1}g} - \dfrac{p_{i}^{N+1, \mathrm{obs}} - p_{i}^{N+1, \mathrm{calc}}}{\left(\sigma_{p_{i}}^{N+1, \mathrm{obs}}\right)^{2}},$$

And the temporal evolution:

$$\mathcal{A}_{i} {S_{S}}_{i} \left(\dfrac{\lambda_{p_{i}}^{n}}{\Delta t^{n-1}} - \dfrac{\lambda_{p_{i}}^{n+1}}{\Delta t^{n}}\right) = \sum_{neigh \;j} \left\lVert \Gamma_{ij} \right\rVert \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{\lambda_{p_{j}}^{n} - \lambda_{p_{i}}^{n}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{\lambda_{p_{j}}^{n+1} - \lambda_{p_{i}}^{n+1}}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert}\Bigg) + \dfrac{\lambda_{h_{i}}^{n}}{\rho_{i}^{n}g} - \dfrac{p_{i}^{n, \mathrm{obs}} - p_{i}^{n, \mathrm{calc}}}{\left(\sigma_{p_{i}}^{n, \mathrm{obs}}\right)^{2}}, \forall n \in [1, N] $$

## Dérivation par $h_{i}^{n}$

$$
\dfrac{\partial \mathcal{L}}{\partial h_{i}^{n}} = - \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\left(\sigma_{h_{i}}^{n, \mathrm{obs}}\right)^{2}} - \lambda_{h_{i}}^{n},\forall n \in [0, N+1] 
$$

Since we want to impose $\dfrac{\partial \mathcal{L}}{\partial p_{i}^{n}} = 0$

$$ \implies \lambda_{h_{i}}^{n} = - \dfrac{h_{i}^{n, \mathrm{obs}} - h_{i}^{n, \mathrm{calc}}}{\left(\sigma_{h_{i}}^{n, \mathrm{obs}}\right)^{2}},\forall n \in [0, N+1] $$

## Derivation by $K_{i}$

$$\dfrac{\partial \mathcal{L}}{\partial K_{i}} = \dfrac{\partial K_{ij}}{\partial K_{i}}\sum_{n=0}^{N} \sum_{neigh \;j} \dfrac{\rho_{ij}^{n}}{\rho_{0}} \dfrac{\left\lVert \Gamma_{ij}  \right\rVert}{\left\lVert \overrightarrow{P_{i}P_{j}} \right\rVert} \Bigg(\alpha_{\mathrm{f}}\left(p_{j}^{n+1} - p_{i}^{n+1}\right) + (1 - \alpha_{\mathrm{f}})\left(p_{j}^{n} - p_{i}^{n}\right)\Bigg)\left(\lambda_{p_{j}}^{n+1} - \lambda_{p_{i}}^{n+1} \right) - \dfrac{K_{i}^{\mathrm{obs}} - K_{i}^{\mathrm{calc}}}{\left(\sigma_{K_{i}}^{\mathrm{obs}}\right)^{2}}$$
