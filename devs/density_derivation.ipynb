# Dérivation de l'équation d'écoulement densitaire

L'équation qui régit la charge hydraulique s'écrit en régime stationaire:

$$
  n = 0 \left\{
    \begin{array}{l}
     \displaystyle \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} K_{ij}\dfrac{h_{j}^{0} - h_{i}^{0}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} - Q_{i}^{n} = 0 \\
     F_{p}^{0} & :& p_{i}^{0} = \dfrac{p_{i}^{0}}{\rho_{i}^{0}g} + z_{i} 
    \end{array}
  \right.
$$


$$
\dfrac{\overline{c}_{2, i}^{n+1} - \overline{c}_{2, i}^{n}}{\Delta t^{n}} = - \dfrac{\overline{c}_{1, i}^{n+1} - \overline{c}_{1, i}^{n}}{\Delta t^{n}} \implies \overline{c}_{2, i}^{n+1} = - \overline{c}_{1, i}^{n+1} + \overline{c}_{1, i}^{n} + \overline{c}_{2, i}^{n}
$$

$$
dM(i, n+1) = \begin{cases}
-\mathrm{min}\left\{-\Delta t^{n} k_{v} A_{s} \overline{c}_{1, i}^{n} \left(1 - \dfrac{c_{1, i}^{n+1}}{K_{s}}\right)c_{2, i}^{n+1}, \overline{c}_{1, i}^{n}, \alpha_{2} c_{2, i}^{n+1}\right\} &, \mathrm{if}\; 0 \leq \dfrac{c_{1, i}^{n+1}}{K_{S}} < 1,  c_{2, i}^{n+1} > 0
\\
0 &, \mathrm{otherwise}
\end{cases}
$$

\begin{equation}
\left\{\begin{array}{l}
\overline{c}_{1, i}^{n+1} = \overline{c}_{1, i}^{n} + dM(i, n+1)
\\
\dfrac{\overline{c}_{2, i}^{n+1} - \overline{c}_{2, i}^{n}}{\Delta t^{n}} = - \dfrac{\overline{c}_{1, i}^{n+1} - \overline{c}_{1, i}^{n}}{\Delta t^{n}} \implies \overline{c}_{2, i}^{n+1} =  \overline{c}_{2, i}^{n} - \alpha_{2}dM(i, n+1)
\end{array}\right.
%\label{eq:chem_law}
\end{equation}

### 1) Initialization of the system (potential saturated flow)

$$
  n=0 \implies \left\{
    \begin{array}{llll}
      1.1)& F_{c}^{0} &:& c^{0}_{i, sp} = \mathrm{cst}, \quad  \forall sp \in [0, N_{sp}-1] 
      \\
      1.2)& F_{\overline{c}}^{0} &:& \overline{c}^{0}_{i, sp} = \mathrm{cst}, \quad \forall sp \in [0, N_{sp}-1] 
      \\
      2) & F_{\rho}^{0} &: & \rho_{i}^{0} = \rho_{0} \left(1 + \displaystyle\sum_{sp} c_{i, sp}^{0} M_{sp} \right) 
      \\
      3)& F_{p}^{0} &:& p_{i}^{0} = \mathrm{cst} \\ % or stationray     
      \\
      4) & F_{h}^{0} & :& h_{i}^{0} = \dfrac{p_{i}^{0}}{\rho_{i}^{0}g} + z_{i} 
      \\
      5) & F_{U}^{0} &: & \overrightarrow{U}_{ij}^{0} = - \dfrac{K_{ij}}{\rho_{0}g} \left( \dfrac{p_{j}^{0} - p_{i}^{0}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}}\right\rVert} \cdot  \overrightarrow{n_{ij}} + \rho_{ij}^{0} g \cdot \overrightarrow{z} \right)
      \end{array}
  \right.
$$

### 2) Reactive transport iterations

$$
\begin{array}{ll}
    1)\; F_{p}: &  \begin{aligned} V_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = &\displaystyle \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) \\
    &+ \alpha_{\mathrm{f}} Q_{i}^{n+1}\rho_{i}^{n}g + (1 - \alpha_{\mathrm{f}}) Q_{i}^{n}\rho_{i}^{n}g \end{aligned}
      \\
      2)\; F_{h}: & h_{i}^{n+1} = \dfrac{p_{i}^{n+1}}{\rho_{i}^{n}g} + z_{i}
      \\
      3)\; F_{U} : & \overrightarrow{U}_{ij}^{n+1} = - \dfrac{K_{ij}}{\rho_{0}g} \left(\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}}\right\rVert} \cdot \overrightarrow{n_{ij}} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \right)
\end{array}
$$

$$
  \forall n \in [0, N_{ts}-1], \left\{
    \begin{array}{ll}
    1) F_{p}: &  \begin{aligned} V_{i} {S_{S}}_{i} \dfrac{p_{i}^{n+1} - p_{i}^{n}}{\Delta t^{n}} = &\displaystyle \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \dfrac{\rho_{ij}^{n}}{\rho_{0}} K_{ij} \Bigg(\alpha_{\mathrm{f}}\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{f}})\dfrac{p_{j}^{n} - p_{i}^{n}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} \rho_{ij}^{n} g \cdot \overrightarrow{z} \cdot \overrightarrow{n_{ij}} \Bigg) \\
    &+ \alpha_{\mathrm{f}} Q_{i}^{n+1}\rho_{i}^{n}g + (1 - \alpha_{\mathrm{f}}) Q_{i}^{n}\rho_{i}^{n}g \end{aligned}
      \\
      2) F_{h}: & h_{i}^{n+1} = \dfrac{p_{i}^{n+1}}{\rho_{i}^{n}g} + z_{i}
      \\
      3) F_{U} : & \overrightarrow{U}_{ij}^{n+1} = - \dfrac{K_{ij}}{\rho_{0}g} \left(\dfrac{p_{j}^{n+1} - p_{i}^{n+1}}{\left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}}\right\rVert} \cdot \overrightarrow{n_{ij}} + \rho_{ij}^{n} g \cdot \overrightarrow{z} \right)
      \\
      4) F_{c}: & \begin{aligned}
 V_{i} \omega_{e,i} \dfrac{c_{i}^{n+1} - c_{i}^{n}}{\Delta t^{n}} & = \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} D_{e, ij}\Bigg(\alpha_{\mathrm{d}}\dfrac{c_{j}^{n+1} - c_{i}^{n+1}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert} + (1 - \alpha_{\mathrm{d}})\dfrac{c_{j}^{n} - c_{i}^{n}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\Bigg)
\\
& \quad - \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \Bigg(\alpha_{\mathrm{a}} c_{ij}^{n+1} \overrightarrow{U_{ij}^{n+1}} + (1 - \alpha_{\mathrm{a}}) c_{ij}^{n} \overrightarrow{U_{ij}^{n}} \Bigg) \cdot \overrightarrow{n_{ij}}
\\
& \quad + \alpha_{\mathrm{a}} \left(\sum_{neigh \;j}  \left(\mathcal{A}_{\Gamma_{ij}} \overrightarrow{U_{ij}^{n+1}} \cdot \overrightarrow{n_{ij}}\right) - Q_{i}^{n+1} \right) c_{i}^{n+1}  
\\
& \quad + (1 - \alpha_{\mathrm{a}}) \left(\sum_{neigh \;j}  \left(\mathcal{A}_{\Gamma_{ij}} \overrightarrow{U_{ij}^{n}} \cdot \overrightarrow{n_{ij}}\right) - Q_{i}^{n} \right) c_{i}^{n}
\\
& \quad + \alpha_{\mathrm{a}} Q_{i, \mathrm{out}}^{n+1} c_{i}^{n+1} + (1 - \alpha_{\mathrm{a}}) Q_{i, \mathrm{out}}^{n} c_{i}^{n} + Q_{i,\mathrm{in}}^{n} c_{i, \mathrm{in}}^{n}
\\
& \quad - V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{n+1} - \overline{c}_{i}^{n}}{\Delta t^{n}}
\end{aligned}
\\
5) F_{\overline{c}_{1}}: & \overline{c}_{1, i}^{n+1} = \overline{c}_{1, i}^{n} + dM(i, n+1)
\\
6) F_{\overline{c}_{2}}: & \overline{c}_{2, i}^{n+1} = \overline{c}_{2, i}^{n} - \alpha_{2}dM(i, n+1)
\\
7) F_{\rho}: & \rho_{i}^{n+1} = \rho_{0} \left(1 + \displaystyle\sum_{sp} c_{i, sp}^{n+1} M_{sp} \right)
\end{array}
\right.
$$
