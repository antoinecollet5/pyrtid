# Notation pour la fonction objective

The objective of the inversion process is to minimize the least-squares misfit function

$$
\mathcal{J}(\mathbf{s}) = \dfrac{1}{2N_{\mathrm{obs}}} \left\lVert \dfrac{d^{\mathrm{obs}} - d^{\mathrm{calc}}}{\sigma_{d^{\mathrm{obs}}}} \right\rVert^{2} = \dfrac{1}{2N_{\mathrm{obs}}} \sum_{k=0}^{N_{\mathrm{obs}}} \left(\dfrac{d_{\mathrm{obs}}^{k} - d_{\mathrm{calc}}(k)}{\sigma_{\mathrm{obs}}^{k}} \right)^{2}
$$

where $s(\bm{x})$ is the model to be determined, which is a function of the spatial coordinates $\mathbf{x} = (x, y, z)$, $d_{\mathrm{obs}}^{k}$ is the $k^{th}$ observed data at observation position $\mathbf{x}^{k}$ and for time $t^{k}$, and $N_{\mathrm{obs}}$ the total number of observed data. The calculated data $d_{\mathrm{calc}}(k)$ are a function of the $\mathbf{s}$ model vector and are defined as:

$$
d_{\mathrm{calc}}(k) = d_{\mathrm{calc}}(\mathbf{x}^k, t^k) = d(\mathbf{x} = \mathbf{x}^k, t=t^k)
$$

For the simplicity of the derivation, the cost function can also be noted as

$$
\mathcal{J}(s) =  \dfrac{1}{2N_{\mathrm{obs}}} \sum_{n=0}^{N_{ts}+1} \sum_{i}^{N_{\Omega}}   \Bigg(\dfrac{d_{i}^{\mathrm{n, obs}} - d_{i}^{\mathrm{n, calc}}}{\sigma_{d_{i}^{\mathrm{n, obs}}}} \Bigg) ^{2} \delta(i, n)
$$

with $N_{ts}$ the number of timesteps in the simulation, $N_{\Omega}$ the number of meshes on the domain, and $\delta(i, n)$ a multivariate dirac distribution defined from the vector of observation times $\mathbf{T}_{\mathrm{obs}}$ and positions $\mathbf{X}_{\mathrm{obs}}$ as:

$$
\delta(i, n) =  \left\{
        \begin{array}{ll}
          1, & \text{if} \;t^{n} \in \mathbf{T}_{\mathrm{obs}}, \text{and} \; \mathbf{x}_{i} \in \mathbf{X}_{\mathrm{obs}}
          \\
          0, & otherwise
        \end{array}
    \right.
$$

Consequently:

$$
\dfrac{\partial\mathcal{J}(s)}{\partial d_{i}^{\mathrm{n, calc}}} = - \dfrac{1}{N_{\mathrm{obs}}} \dfrac{d_{i}^{\mathrm{n, obs}} - d_{i}^{\mathrm{n, calc}}}{\sigma_{d_{i}^{\mathrm{n, obs}}}} \delta(i, n)
$$
