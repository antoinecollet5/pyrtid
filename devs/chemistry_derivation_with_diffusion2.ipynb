L'équation de dissolution d'un mineral traceur et de l'évolution des concentrations s'écrit:

- En version explicite

$$
  \left\{
    \begin{array}{l}
      \overline{c}_{i}^{2n+1} = \overline{c}_{i}^{2n}
      \\
      c_{i}^{2n+1} = 1.05 * D_{i} c_{i}^{2n}
      \\
      V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}} = V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right)
      \\
      V_{i} \omega_{e, i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n+2}} = - V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}}
    \end{array}
  \right.
$$

with the fix point criterion: $$ \left(1 - \dfrac{c_{i}^{2n+2}}{c_{i}^{2n}} \right) < \epsilon $$

$$ J = \dfrac{1}{2} \sum_{n} \sum_{i} \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{\sigma_{c_{i}^{n, obs}}}\right)^{2} - \langle \lambda_{c_{i}^{n+1}}, g_{c} \rangle - \langle \lambda_{\overline{c}_{i}^{n+1}}, g_{\overline{c}} \rangle $$

Derivation par $\overline{c}_{i}^{2n}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n}} \langle \lambda_{\overline{c}_{i}}^{2n+1}, \overline{c}_{i}^{2n+1} - \overline{c}_{i}^{2n} \rangle$$

$$ = \lambda_{\overline{c}_{i}}^{2n} - \lambda_{\overline{c}_{i}}^{2n+1}$$

Derivation par $\overline{c}_{i}^{2n+1}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n+1}} \langle \lambda_{\overline{c}_{i}}^{2n+2}, V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}} - V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \rangle$$

$$ = \lambda_{\overline{c}_{i}}^{2n+2} \Bigg( \dfrac{-V_{i} \omega_{e, i}}{\Delta t^{2n+2}} - V_{i} \omega_{e, i} k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \Bigg) + \lambda_{\overline{c}_{i}}^{2n+1} \Bigg( \dfrac{V_{i} \omega_{e, i}}{\Delta t^{n}}\Bigg) $$

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{2n+1}} \langle \lambda_{c_{i}}^{2n+2}, V_{i} \omega_{e, i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n+2}} + V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}} \rangle$$

$$ = V_{i} \omega_{e, i} \left( \dfrac{-\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} + \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+1}} \right)$$

On obtient alors une expression explicite pour les teneurs adjointes:

$$ \lambda_{\overline{c}_{i}}^{2n+1} = \Bigg( \lambda_{\overline{c}_{i}}^{2n+2} \left( \dfrac{1}{\Delta t^{2n+2}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right) - \left( \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+1}} - \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} \right) \Bigg) \Delta t^{2n+1}$$

2) Derivation par $c_{i}^{2n}$:


$$ \dfrac{\partial}{\partial c_{i}^{2n}} \langle \lambda_{c_{i}}^{2n+1}, V_{i} \omega_{e, i} \dfrac{c_{i}^{2n+1} - c_{i}^{2n}}{\Delta t^{2n+2}} - \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \big(D_{e, ij}\dfrac{c_{j}^{2n+1} - c_{i}^{2n+1}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\big) \rangle$$

$$= V_{i} \omega_{e, i} \Big( \dfrac{\lambda_{c_{i}}^{2n}}{\Delta t^{2n+1}} - \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+2}} \Big) - \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \left( D_{e, ij}\dfrac{\lambda_{c_{j}}^{2n} - \lambda_{c_{i}}^{2n}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\right) $$

$$ \dfrac{\partial}{\partial c_{i}^{2n}} \langle \lambda_{c_{i}}^{2n+1}, V_{i} \omega_{e, i} c_{i}^{2n+1} - V_{i} \omega_{e, i} 1.05 * D_{i} * c_{i}^{2n} \rangle$$

$$ \implies V_{i} \omega_{e, i} \lambda_{c_{i}}^{2n} = V_{i} \omega_{e, i} 1.05 D_{i} \lambda_{c_{i}}^{2n+1} $$

3) Derivation par $c_{i}^{2n+1}$:

$$ \dfrac{\partial}{\partial c_{i}^{2n+1}} \langle \lambda_{\overline{c}_{i}}^{2n+2}, V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}} - V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \rangle$$

$$ = V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \dfrac{\lambda_{\overline{c}_{i}}^{2n+2}}{K_{s}} $$

$$ \dfrac{\partial}{\partial c_{i}^{2n+1}} \langle \lambda_{c_{i}}^{2n+2}, V_{i} \omega_{e, i} \dfrac{c_{i}^{2n+2} - c_{i}^{2n+1}}{\Delta t^{2n+2}} + V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{2n+2} - \overline{c}_{i}^{2n+1}}{\Delta t^{2n+2}} \rangle$$

$$ = V_{i} \omega_{e, i}  \left( \dfrac{- \lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} + \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+1}} \right)$$

$$ \dfrac{\partial}{\partial c_{i}^{2n}} \dfrac{1}{2} \sum_{n} \sum_{i} \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{\sigma_{c_{i}^{2n, obs}}}\right)^{2}$$

$$ = - \left( \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{\sigma_{c_{i}^{2n, obs}}}^{2}}\right)$$

Gathering the three previous, it yields

$$ - V_{i} \omega_{e, i}  \left( \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+1}} \right) - V_{i} \omega_{e, i} \left( \dfrac{- \lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} \right)  - \left( \dfrac{c_{i}^{2n+2, obs} - c_{i}^{2n+2, calc}}{{\sigma_{c_{i}^{2n=2, obs}}}^{2}}\right) - V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{2n+1} \dfrac{\lambda_{\overline{c}_{i}}^{2n+2}}{K_{s}} = 0 $$

$$ \implies \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} -  \lambda_{\overline{c}_{i}}^{2n+2} \dfrac{k_{v} A_{s} \overline{c}_{i}^{2n+1}}{K_{s}}  - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{V_{i} \omega_{e, i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n+1} $$

Derivation du point fixe:

$$ \dfrac{\partial}{\partial c_{i}^{2n}} \langle \lambda_{c_{i}}^{2n+2}, \left(1 - \dfrac{c_{i}^{2n+2}}{c_{i}^{2n}} \right) - \epsilon \rangle$$


$$ \implies \dfrac{\lambda_{c_{i}}^{2n+2}}{{c_{i}^{2n}}^{2}} - \lambda_{c_{i}}^{2n} < 0 $$

Dérivation par rapport à $\overline{c}_{i}^{0}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \langle \lambda_{\overline{c}_{i}}^{1}, \overline{c}_{i}^{1} - \overline{c}_{i}^{0} \rangle = \lambda_{\overline{c}_{i}}^{0} - \lambda_{\overline{c}_{i}}^{1}$$

Dérivation par rapport à $\overline{c}_{i}^{0}$:

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \langle \lambda_{\overline{c}_{i}^{1}}, V_{i} \omega{i} \dfrac{\overline{c}_{i}^{1} - \overline{c}_{i}^{0}}{\Delta t^{0}} - V_{i} \omega_{e, i} k_{v} A_{s} \overline{c}_{i}^{0} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \rangle$$


$$ = V_{i} \omega_{e, i} \lambda_{\overline{c}_{i}^{1}} \left(\dfrac{- 1}{\Delta t^{0}} - k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right)$$

$$ \dfrac{\partial}{\partial \overline{c}_{i}^{0}} \langle \lambda_{c_{i}^{1}}, V_{i} \omega_{e, i} \dfrac{c_{i}^{1} - c_{i}^{0}}{\Delta t^{1}} + V_{i} \omega_{e, i} \dfrac{\overline{c}_{i}^{1} - \overline{c}_{i}^{0}}{\Delta t^{1}} \rangle$$


$$ = - V_{i} \omega_{e, i} \dfrac{\lambda_{c_{i}^{1}}}{\Delta t^{1}}$$

$$\implies \dfrac{d J}{d \overline{c}_{i}^{0}} = V_{i} \omega_{e, i} \lambda_{\overline{c}_{i}^{1}} \left(\dfrac{1}{\Delta t^{0}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{0}}{K_{s}}\right) \right) + V_{i} \omega_{e, i} \dfrac{\lambda_{c_{i}^{1}}}{\Delta t^{1}}$$


$$
  \left\{
    \begin{array}{l}
      \lambda_{c_{i}}^{2n+1} = \left( \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} -  \lambda_{\overline{c}_{i}}^{2n+2} \dfrac{k_{v} A_{s} \overline{c}_{i}^{2n+1}}{K_{s}}  - \dfrac{c_{i}^{2n, obs} - c_{i}^{2n, calc}}{{V_{i} \omega_{e, i} \sigma_{c_{i}^{2n, obs}}}^{2}} \right) \Delta t^{2n+1}
      \\
      \lambda_{\overline{c}_{i}}^{2n+1} = \Bigg( \lambda_{\overline{c}_{i}}^{2n+2} \left( \dfrac{1}{\Delta t^{2n+2}} + k_{v} A_{s} \left( 1 - \dfrac{c_{i}^{2n+1}}{K_{s}}\right) \right) - \left( \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+1}} - \dfrac{\lambda_{c_{i}}^{2n+2}}{\Delta t^{2n+2}} \right) \Bigg) \Delta t^{2n+1}
      \\
      V_{i} \omega_{e, i} \Big( \dfrac{\lambda_{c_{i}}^{2n}}{\Delta t^{2n+1}} - \dfrac{\lambda_{c_{i}}^{2n+1}}{\Delta t^{2n+2}} \Big) - \sum_{neigh \;j} \mathcal{A}_{\Gamma_{ij}} \left( D_{e, ij}\dfrac{\lambda_{c_{j}}^{2n} - \lambda_{c_{i}}^{2n}}{ \left\lVert \overrightarrow{\mathrm{P}_{i}\mathrm{P}_{j}} \right\rVert}\right)
      \\
      \lambda_{\overline{c}_{i}}^{2n} = \lambda_{\overline{c}_{i}}^{2n+1}
    \end{array}
  \right.
$$

- Diffusion gradient

$$ \dfrac{\partial}{\partial D_{i}} \langle \lambda_{c_{i}}^{2n+1}, V_{i} \omega_{e, i} c_{i}^{2n+1} - V_{i} \omega_{e, i} (1 + D_{i}) c_{i}^{2n} \rangle$$

$$ = \lambda_{c_{i}}^{2n+1} V_{i} \omega_{e, i} c_{i}^{2n} $$

```python
from typing import List
import numpy as np
import matplotlib.pyplot as plt
```

```python
class GeochemicalParameters:
    """
    Class defining the geocgemical parameters used in the simulation.

    Attributes
    ----------
    conc: float, optional
        Initial tracer concentration in the grid in molal. The default is 0.0.
    grade: float, optional
        Default mineral grade in the grid in mol/kg (kg of water).
        The default is 0.0.
    kv: float, optional
        The kinetic rate of the mineral in [mol/m2/s]. The default is -6.9e-9.
    As: float, optional
        Specific area in [m2/mol]. The default is 13.5.
    Ks: float, optional
        Solubility constant (no unit). The default is 6.3e-4.
    """

    def __init__(
        self,
        conc: float = 0.0,
        grade: float = 0.0,
        kv: float = -6.9e-9,
        As: float = 13.5,
        Ks: float = 6.3e-4,
    ) -> None:
        """Initialize the instance."""
        self.conc: float = conc
        self.grade: float = grade
        self.kv: float = kv
        self.As: float = As
        self.Ks: float = Ks


class Geometry:
    """
    Class defining the grid geometry used in the simulation.

    Attributes
    ----------
    nx : int
        Number of voxels along the x axis.
    ny : int
        Number of voxels along the y axis.
    dx : float
        Voxel dimension along the x axis.
    dy : float
        Voxel dimension along the y axis.
    dz : float
        Voxel dimension along the z axis. This is only taken into account for
        the voxel volume computation.
    """

    def __init__(
        self,
        nx: int,
        ny: int,
        dx: float,
        dy: float,
        dz: float = 1.0,
    ) -> None:
        """Initialize the class instance."""
        self._nx = 1
        self._ny = 1
        self.nx = int(nx)
        self.ny = int(ny)
        self.dx: float = dx
        self.dy: float = dy
        self.dz: float = dz

    @property
    def nx(self) -> int:
        """Return the number of meshes along the x axis."""
        return self._nx

    @nx.setter
    def nx(self, value: int) -> None:
        if value < 1:
            raise (ValueError("nx should be > 1!)"))
        self._nx = value

    @property
    def ny(self) -> int:
        """Return the number of meshes along the y axis."""
        return self._ny

    @ny.setter
    def ny(self, value: int) -> None:
        self._ny = value

    @property
    def mesh_area(self) -> float:
        """Return the area of a voxel in m2."""
        return self.dx * self.dy

    @property
    def mesh_volume(self) -> float:
        """Return the volume of a voxel in m3."""
        return self.mesh_area * self.dz


class TimeParameters:
    """
    Class defining the time parameters used in the simulation.

    It also handles the variable timestep.

    Attributes
    ----------
    nt : int
        Number of timesteps in the simulation.
    dt : float
        Current timestep in seconds.
    dt_init : float
        Initial timestep in seconds.
    dt_min : float
        Minimum timestep in seconds.
    dt_max : float
        Maximum timestep in seconds.
    ldt: List[float]
        List of successive timesteps (in seconds) used in the forward modelling.
    """

    __slots__ = ["nt", "dt", "dt_init", "dt_min", "dt_max", "ldt"]

    def __init__(self, nt: int, dt_init: float, dt_min: float, dt_max: float) -> None:
        """Initialize the instance."""
        self.nt: int = int(nt)
        self.dt_init: float = dt_init
        self.dt: float = dt_init
        self.dt_min: float = dt_min
        self.dt_max: float = dt_max
        self.ldt: List[float] = []

    @property
    def duration(self) -> float:
        """Simulation duration in seconds."""
        return self.nt * self.dt

    def reset_dt(self) -> None:
        """Empty the list of timesteps and set dt to its initial value."""
        self.dt = self.dt_init
        self.ldt = []

    def update_dt(self, n_iter: int) -> None:
        """
        Update the timestep.

        Parameters
        ----------
        n_iter: int
            Number of iterations required to solve the current timestep.
        """
        if n_iter < 20:
            # increase dt by 5%
            self.dt *= 1.05
        else:
            # decrease dt by 30%
            self.dt *= 0.7
        # Ensure timebounds
        if self.dt < self.dt_min:
            self.dt = self.dt_min
        if self.dt > self.dt_max:
            self.dt = self.dt_max
        self.ldt.append(self.dt)


class TransportParameters:
    """
    Class defining the transport parameters used in the simulation.

    Attributes
    ----------
    diffusion: float, optional
        Default diffusion coefficient in the grid in [m2/s]. The default is 1e-4 m2/s.
    porosity: float, optional
        Default porosity in the grid Should be a number between 0 and 1.
        The default is 1.0.
    crank_nicolson: float
        The Crank-Nicholson parameter allows to set the temporal resolution
        scheme to explicit, fully implicit or somewhere in between these two
        extremes. The value must be comprised between 0.0 and 1.0, 0.0 being a
        full explicit scheme and 1.0 fully implicit. The default is 0.5.
    tolerance: float, optional
        The tolerance on the transport. The default is 1e-8.
    is_numerical_acceleration: bool, optional
        Whether to use the concentration values at the previous iteration as an initial
        guess for gmres (x0) rather than using a random vector. This should increase the
        speed of resolution. The default is False.
    """

    def __init__(
        self,
        diffusion: float = 1e-4,
        porosity: float = 1.0,
        crank_nicolson_advection: float = 0.5,
        crank_nicolson_diffusion: float = 1.0,
        tolerance: float = 1e-8,
        is_numerical_acceleration: bool = False,
    ) -> None:
        """Initialize the instance."""
        self.diffusion: float = diffusion
        self.porosity: float = porosity
        self.crank_nicolson_advection: float = crank_nicolson_advection
        self.crank_nicolson_diffusion: float = crank_nicolson_diffusion
        self.tolerance: float = tolerance
        self.is_numerical_acceleration: bool = is_numerical_acceleration
```

```python
nx = 20  # number of voxels along the x axis
ny = 1  # number of voxels along the y axis
dx = 1.0  # 2.98  # voxel dimension along the x axis
dy = 1.0  # 3.7  # voxel dimension along the y axis
dz = 1

geom = Geometry(nx, ny, dx, dy, dz)

time_params = TimeParameters(
    nt=50,  # number of time steps
    dt_init=500000.0,  # timestep in seconds
    dt_min=500.0,
    dt_max=5000000.0,
)

# Chemistry parameters
c0 = 1e-4  # general initial concentration [molal]
M0 = 0.001  # mineral grade [mol/kg] -> kg of water
M1 = 0.0005  # mineral grade [mol/kg] -> kg of water
M1 = np.abs(0.001 * (0.5 + 0.5 * np.sin(np.arange(nx) / 2))).reshape(20, -1)

M0 = 0.0
M1 = 0.0

kv = -6.9e-9  # kinetic rate,       [mol/m2/s]
moleweight = 270.0  # molar weight [g/mol]
surface = 500  # cm2/g
As = moleweight * surface / 1e4  # specific area [m2/mol]
logK = 3.2
Ks = 1.0 / pow(10, logK)  # solubility constant [no unit]
porosity = 1.0

diffusion_ref = np.ones((nx, 1)) * 1e-2
diffusion_est = np.abs(0.01 * (0.5 + 0.5 * np.sin(np.arange(nx) / 2))).reshape(20, -1)

gch_params_ref = GeochemicalParameters(c0, M0, kv, As, Ks)

gch_params_est = GeochemicalParameters(c0, M1, kv, As, Ks)

tr_params_ref = TransportParameters(diffusion=diffusion_ref, porosity=porosity)
tr_params_est = TransportParameters(diffusion=diffusion_est, porosity=porosity)
```

```python
class TransportModel:
    def __init__(
        self, nt: int, geochem: GeochemicalParameters, tr_params: TransportParameters
    ) -> None:
        self.conc = np.zeros((nx, ny, nt + 1))
        self.conc_post_tr = np.zeros((nx, ny, nt + 1))
        self.conc[:, :, 0] = geochem.conc
        self.grade = np.zeros((nx, ny, nt + 1))
        self.grade_post_tr = np.zeros((nx, ny, nt + 1))
        self.grade[:, :, 0] = geochem.grade
        self.porosity = np.zeros((nx, ny))
        self.porosity[:, :] = tr_params.porosity
        self.diffusion = np.zeros((nx, ny))
        self.diffusion[:, :] = tr_params.diffusion


class AdjointTransportModel:
    def __init__(self, nt: int) -> None:
        self.a_conc = np.zeros((nx, ny, nt + 1))
        self.a_conc_post_gch = np.zeros((nx, ny, nt + 1))
        self.a_grade = np.zeros((nx, ny, nt + 1))
        self.a_grade_post_gch = np.zeros((nx, ny, nt + 1))
```

```python
def solve_transport(
    tr_model: TransportModel,
    gch_params: GeochemicalParameters,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the geochemistry part."""

    # Need to take into account boundary conditions:
    # And then the reactive (chemistry) contribution with the updated conc

    # Note: the concentration has been modified by the transport so we need to use the
    # updated one.
    tr_model.grade[:, :, time_index] = tr_model.grade[:, :, time_index - 1]
    tr_model.conc[:, :, time_index] = (1.0 + tr_model.diffusion) * tr_model.conc[
        :, :, time_index - 1
    ]


def solve_geochem_explicit(
    tr_model: TransportModel,
    gch_params: GeochemicalParameters,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the geochemistry part."""

    # Need to take into account boundary conditions:
    # And then the reactive (chemistry) contribution with the updated conc

    # Note: the concentration has been modified by the transport so we need to use the
    # updated one.
    m0 = tr_model.grade[:, :, time_index - 1]

    dMdt = (
        gch_params.kv
        * gch_params.As
        * m0
        * (1.0 - tr_model.conc[:, :, time_index - 1] / gch_params.Ks)
    )

    tr_model.grade[:, :, time_index] = m0 + dMdt * time_params.dt
    tr_model.conc[:, :, time_index] = (
        tr_model.conc_post_tr[:, :, time_index] - dMdt * time_params.dt
    )
```

```python
tr_model_ref = TransportModel(time_params.nt, gch_params_ref, tr_params_ref)

tr_model_est = TransportModel(time_params.nt, gch_params_est, tr_params_est)


dt_list = [0.0]
time_params.reset_dt()
for it in range(1, time_params.nt + 1):
    time_params.dt
    time_params.update_dt(1)  # we consider only one iteration.
    solve_transport(tr_model_ref, gch_params_ref, time_params, it)
    # solve_geochem_explicit(tr_model_ref, gch_params_ref, time_params, it)

time_params.reset_dt()
for it in range(1, time_params.nt + 1):
    time_params.update_dt(1)  # we consider only one iteration.
    solve_transport(tr_model_est, gch_params_est, time_params, it)
    # solve_geochem_explicit(tr_model_est, gch_params_est, time_params, it)
```

  \right.
$$

```python
def solve_adj_geochem_explicit(
    tr_model: TransportModel,
    a_tr_model: AdjointTransportModel,
    gch_params: GeochemicalParameters,
    geometry: Geometry,
    a_sources,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    """Compute the geochemistry part."""

    # Adjoint variable at the previous timestep going backward (timestep n+1)
    if time_index + 1 == a_tr_model.a_grade.shape[-1]:
        # Handle the first time step in the adjoint (= last timestep in the forward)
        am_prev = 0.0  # initially at zero for the first timestep
        ac_prev = 0.0  # initially at zero for the first timestep
        dt_prev = time_params.ldt[time_index - 1]
    else:
        am_prev = a_tr_model.a_grade[:, :, time_index + 1]
        ac_prev = a_tr_model.a_conc[:, :, time_index + 1]
        dt_prev = time_params.ldt[time_index]

    # Forward variable at the current timestep going backward
    # (timestep n -> we want to predict)
    m_cur = tr_model.grade[:, :, time_index]
    c_cur = tr_model.conc[:, :, time_index]
    dt_cur = time_params.ldt[time_index - 1]

    # Update concentration (with adjoint sources)
    a_tr_model.a_conc_post_gch[:, :, time_index] = (
        # ac_prev / dt_prev
        -am_prev * gch_params.kv * gch_params.As * m_cur / gch_params.Ks
        - a_sources[:, :, time_index] / geometry.mesh_area / tr_model.porosity
    )  # * dt_cur

    # a_tr_model.a_conc_post_gch[:, :, time_index] = ac_prev - a_sources[:, :, time_index] / geometry.mesh_area / tr_model.porosity

    # Update mineral value
    # a_tr_model.a_grade_post_gch[:, :, time_index] = (
    #     am_prev
    #     * (1 / dt_prev + gch_params.kv * gch_params.As * (1.0 - c_cur / gch_params.Ks))
    #     - (a_tr_model.a_conc_post_gch[:, :, time_index] / dt_cur - ac_prev / dt_prev)
    # ) * dt_cur
```

```python
def solve_adj_transport(
    tr_model: TransportModel,
    a_tr_model: AdjointTransportModel,
    gch_params: GeochemicalParameters,
    geometry: Geometry,
    a_sources: float,
    time_params: TimeParameters,
    time_index: int,
) -> None:
    try:
        ac = a_tr_model.a_conc[:, :, time_index + 1]
    except IndexError:
        ac = 0.0
    a_tr_model.a_conc[:, :, time_index] = (1.0 + tr_model.diffusion) * ac - a_sources[
        :, :, time_index
    ]
    a_tr_model.a_grade[:, :, time_index] = a_tr_model.a_grade_post_gch[:, :, time_index]
```
