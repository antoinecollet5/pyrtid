```python
import matplotlib.pyplot as plt
import numpy as np
import pyrtid.inverse as dminv
```

```python
pcd = dminv.Uniform2Gaussian(ud_lbound=-3, ud_ubound=5.0, gd_mu=2.0, gd_std=12.78)

# dist = np.random.default_rng(256).uniform(pcd.ud_lbound, pcd.ud_ubound, size=10000)
dist_g = np.random.default_rng(256).normal(loc=pcd.gd_mu, scale=pcd.gd_std, size=10000)

plt.hist(dist_g, bins=20, density=True)
```

```python
dist_u = pcd.backtransform(dist_g)
plt.hist(dist_u, bins=20, density=True)
```

```python
plt.hist(pcd(dist_u), bins=20, density=True)
```

```python
np.testing.assert_allclose(pcd(dist_u), dist_g)
```
