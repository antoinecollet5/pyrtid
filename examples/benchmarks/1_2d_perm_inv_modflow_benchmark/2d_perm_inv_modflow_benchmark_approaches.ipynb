# pyPCGA modflow inversion example

## Crosswell tomography

We test:
- true
- no regularization (no_reg)
- gradient smoothing (gs)
- tikonov adaptative (tka)
- gradient smoothing + tikhonov adaptative (gstka)
- tv
- sigmoid
- sigmoid + tv

- Adapted from Application 1 of Lee and Kitanids [WRR 2014] 

(https://agupubs.onlinelibrary.wiley.com/doi/full/10.1002/2014WR015483)

***

> We first generated 870 head observations using crosswell pumping tests (ln 9) from the reference log K field (ln 11) with USGS MODFLOW Flopy python interface.

> In specific, we extracted groundwater at a well and observed 29 stready-state hydraulic heads at the remaining monitoring wells. We repeated these crosswell pumping tests for all 30 well locations (ln 11).  

> Then add the noise (std = 0.5) to observations. Using observations with MODFLOW, can we reconstruct the reference field (ln 11)?

Note the example is not exactly same as Application 1 in Lee and Kitanidis [2014]

Also, the number of MODFLOW simulations is higher than what they reported because we turned on Levenberg-Marquardt and Line Search options in this example. Those additional optimization routines increases the inversion time, but usually results in a very accurate inverse solution even for highly nonlinear problems.

Note: in the demonstrator, we can adjust the permeability (m/s), so we need to convert these values to permeability ones



```python
def log_transmitivity_to_permeability(x: NDArrayFloat) -> NDArrayFloat:
    """Convert a transmitivity in ln[m2/d] to a permeability in [m/s]."""
    return np.exp(x) / 3600 / 24


def permeability_to_log_transmitivity(x: NDArrayFloat) -> NDArrayFloat:
    """Convert a permeability in [m/s] to a transmitivity in ln[m2/d]."""
    return np.log(x * 3600 * 24)


logT_reference: NDArrayFloat = np.loadtxt("../../data/reference_logK.txt")
perm_reference = log_transmitivity_to_permeability(logT_reference)
perm_reference_2d = perm_reference.reshape(ny, nx).T
logT_reference_3d = logT_reference.reshape(nz, ny, nx)
logT_reference_2d = logT_reference_3d[0, :, :].T

plotter = ngp.NestedGridPlotter(
    plt.figure(constrained_layout=True, figsize=(15, 5)),
    builder=ngp.SubplotsMosaicBuilder(
        mosaic=[["ax1-1", "ax1-2"]], sharey=True, sharex=True
    ),
)

# Plot the permeability
ngp.multi_imshow(
    axes=[plotter.get_axis("ax1-1")],
    fig=plotter.fig,
    data={"reference logK": logT_reference_2d},
    cbar_title="Transmitivity field (logK), $ln[m2/d]$",
    imshow_kwargs={"cmap": plt.get_cmap("jet"), "extent": [0.0, nx * dx, 0.0, ny * dy]},
    xlabel="X [m]",
    ylabel="Y [m]",
)

ngp.multi_imshow(
    axes=[plotter.get_axis("ax1-2")],
    fig=plotter.fig,
    data={"reference permeability": perm_reference_2d},
    cbar_title="Permeability field $[m/s]$",
    imshow_kwargs={
        "cmap": plt.get_cmap("jet"),
        "norm": LogNorm(vmin=1e-5, vmax=1e-3),
        "extent": [0.0, nx * dx, 0.0, ny * dy],
    },
    xlabel="X [m]",
    ylabel="Y [m]",
)
```

# Forward Model with PyRTID

- Creation of a base model

```python
duration_in_days = 31
duration_in_s = duration_in_days * 3600 * 24
dt_init = 3600 * 12  # timestep in seconds
dt_min = dt_init
dt_max = dt_init

# Simulation on 31 days. We use a 4h timestep.
time_params = dmfwd.TimeParameters(
    duration=duration_in_s, dt_init=dt_init, dt_max=dt_max, dt_min=dt_min
)

geometry = dmfwd.Geometry(nx=nx, ny=ny, dx=dx, dy=dy)
fl_params = dmfwd.FlowParameters(
    permeability=8.0e-5,
    storage_coefficient=1e-6,
    regime=dmfwd.FlowRegime.STATIONARY,
    crank_nicolson=1.0,
)
tr_params = dmfwd.TransportParameters(is_numerical_acceleration=True, is_skip_rt=True)

base_model = dmfwd.ForwardModel(geometry, time_params, fl_params, tr_params)
```

- Add boundary conditions and regional recharge

```python
# Add a regional recharge -> 1000 m3/d = 1000 / 3600 / 24 m3/s
flw_reg = 1000.0 / 3600 / 24
recharge_source = dmfwd.SourceTerm(
    "aquifer_recharge",
    node_ids=np.arange(nx * ny),  # -> on the full domain
    times=np.arange(time_params.nt) * time_params.dt,
    flowrates=np.repeat(flw_reg, time_params.nt),
    concentrations=np.repeat(0.0, time_params.nt),
)
base_model.add_src_term(recharge_source)

# Boundary conditions
# Left border (x = 0) at 120 m
base_model.add_boundary_conditions(dmfwd.ConstantHead(span=(slice(0, 1), slice(None))))
base_model.fl_model.set_initial_head(120.0, span=(slice(0, 1), slice(None)))
# Right border (x = xmax) at 110 m
base_model.add_boundary_conditions(
    dmfwd.ConstantHead(span=(slice(nx - 1, nx), slice(None)))
)
base_model.fl_model.set_initial_head(110.0, span=(slice(nx - 1, nx), slice(None)))
```

- Create the source terms for the pumping: one pumping test per day at -25 m3/h

```python
flowrates = np.array([-25.0 / 3600, 0.0])
for day in range(1, int(duration_in_days)):
    ix, iy = obs_locs[day - 1]
    source_term = dmfwd.SourceTerm(
        "well pumping test # {day}",
        node_ids=np.array(
            indices_to_node_number(ix=ix, nx=nx, iy=iy, ny=ny)
        ),  # -> on the full domain
        times=np.array([day * 3600 * 24, (day + 0.5) * 3600 * 24]),
        flowrates=flowrates,
        concentrations=np.zeros((2,)),
    )
    base_model.add_src_term(source_term)
```

- Create two models with different permeabilities

```python
# copy the base model
model_reference = copy.deepcopy(base_model)
# update permeability
model_reference.fl_model.permeability = perm_reference_2d
# solve the flow problem
dmfwd.ForwardSolver(model_reference).solve(is_verbose=True)
```

```python
# copy the base model
model_estimate = copy.deepcopy(base_model)
# update permeability
model_estimate.fl_model.permeability[:, :] = log_transmitivity_to_permeability(1.0)
# solve the flow problem
dmfwd.ForwardSolver(model_estimate).solve()
```

```python
noise_std = 0.5  # This is an absolute value


def get_white_noise(shape: Iterable[int]) -> NDArrayFloat:
    """
    Return the input with some added white noise.

    Note
    ----
    The parameters are hardcoded to be consistent in the notebook.
    Change the function directly.
    """
    mean_noise = 0.0  # mean
    return np.random.default_rng(2021).normal(
        mean_noise, noise_std, size=np.array(shape)
    )


obs_noise: NDArrayFloat = get_white_noise((len(obs_locs), 29))
```

# Inversion

- Define observables

```python
observables: List[dminv.Observable] = []
for count, (ix, iy) in enumerate(obs_locs):
    vals = model_reference.fl_model.head[ix, iy, 3::2]  # 30 measures from day 1 to 30
    times = model_reference.time_params.times[3::2]
    # Remove the pumping day
    vals = np.delete(vals, count)
    times = np.delete(times, count)

    observables.append(
        dminv.Observable(
            state_variable=dminv.StateVariable.HEAD,
            node_indices=np.array(indices_to_node_number(ix=ix, nx=nx, iy=iy, ny=ny)),
            times=times,
            values=vals + obs_noise[count],
            uncertainties=noise_std,
        )
    )
```

- Create a simulation for inversion based

- true
- no regularization (no_reg)
- gradient smoothing (gs)
- tikonov adaptative (tka)
- gradient smoothing + tikhonov adaptative (gstka)
- tv (tva)
- sigmoid (sg)
- sigmoid + tv (gstv)

```python
solver_config = dminv.LBFGSBSolverConfig(
    maxfun=3,
    maxiter=30,
    ftol=1e-7,
    gtol=1e-7,
    is_check_gradient=False,
    is_adj_numerical_acceleration=True,
    afpi_eps=1e-10,
)
```

```python
import numpy as np
from scipy.optimize import LbfgsInvHessProduct
from scipy.sparse.linalg import LinearOperator, gmres
from pyrtid.inverse.preconditioner import NoTransform


from pyrtid.utils.types import NDArrayFloat
from scipy._lib._util import check_random_state  # To handle random_state
from pyrtid.utils import extract_hess_inv_diag

from typing import Optional, Union
from pyesmda.utils import get_ensemble_variance
from pyrtid.inverse.regularization.covariances import (
    CovarianceMatrix,
    EigenFactorizedCovarianceMatrix,
    eigen_factorize_cov_mat,
    CallBack,
)


def recover_matrix_from_operator(lin_op: LinearOperator, n: int) -> csc_array:
    """Recover matrix from linear operator using unit vectors (slow)."""
    return sp.sparse.vstack(
        [
            sp.sparse.csc_array(lin_op(unit_vector.toarray().T))
            for unit_vector in sp.sparse.eye(n).tocsc()
        ]
    )


class HessInvProductCovarianceMatrix(CovarianceMatrix):
    """Represents a Inverse Hessian Product (LBFGSB) covariance matrix."""

    def __init__(
        self,
        hess_inv: LbfgsInvHessProduct,
    ) -> None:
        """
        Initialize the instance.

        Parameters
        ----------
        hess_inv : LbfgsInvHessProduct
            Linear operator for the L-BFGS approximate inverse Hessian.
            This operator computes the product of a vector with the approximate inverse
            of the Hessian of the objective function, using the L-BFGS limited memory
            approximation to the inverse Hessian, accumulated during the optimization.
            We rely on the property that when L-BFGS has converged, the inverse Hessian
            is an approximation of the covariance matrix a posteriori.
        """
        self.hess_inv: LbfgsInvHessProduct = hess_inv
        super().__init__((hess_inv.shape[0], hess_inv.shape[0]))

    def _matvec(self, x: NDArrayFloat) -> NDArrayFloat:
        """Return the covariance matrix times the vector x."""
        return self.hess_inv(x)

    def _rmatvec(self, x: NDArrayFloat) -> NDArrayFloat:
        """Return the covariance matrix conjugate transpose times the vector x."""
        return self.hess_inv(x)

    def solve(
        self, b: NDArrayFloat, tol: float = 1e-12, maxiter: int = 1000
    ) -> NDArrayFloat:
        """Solve Ax = b, with A, the current covariance matrix instance."""
        residual = CallBack()
        x, info = gmres(
            self,
            b,
            tol=tol,
            maxiter=maxiter,
            callback=residual,
            M=None,
            atol=0.0,
            callback_type="legacy",
        )
        self.solvmatvecs += residual.itercount
        return x

    def get_diagonal(self) -> NDArrayFloat:
        """Return the diagonal entries of the matrix (variances)."""
        return extract_hess_inv_diag(self.hess_inv)

    def get_trace(self) -> float:
        """Return the trace (sum of the diagonal) of the covariance matrix."""
        return float(np.sum(self.get_diagonal()))


def get_sparse_eigen_factor_hess(
    hess_inv: LbfgsInvHessProduct,
    n_pc: int = 200,
    random_state: Optional[
        Union[int, np.random.Generator, np.random.RandomState]
    ] = None,
) -> csc_array:
    """
    Factorize as AA^T the linear operator for the L-BFGS approximate inverse Hessian.

    A is a sparse square matrix with size (n, n) with only the n_pc first columns
    non null.

    Parameters
    ----------
    hess_inv : LbfgsInvHessProduct
        Linear operator for the L-BFGS approximate inverse Hessian.
        This operator computes the product of a vector with the approximate inverse
        of the Hessian of the objective function, using the L-BFGS limited memory
        approximation to the inverse Hessian, accumulated during the optimization.
    n_pc : int
        Number of principal component kept (eigen vectors and values). The default
        is 200.
    random_state : Optional[Union[int, np.random.Generator, np.random.RandomState]]
        Pseudorandom number generator state used to generate resamples.
        If `random_state` is ``None`` (or `np.random`), the
        `numpy.random.RandomState` singleton is used.
        If `random_state` is an int, a new ``RandomState`` instance is used,
        seeded with `random_state`.
        If `random_state` is already a ``Generator`` or ``RandomState``
        instance then that instance is used. The default is None

    Returns
    -------
    csc_array
        Sparse square factor A with size (n, n) and only the n_pc first columns non
        null.
    """
    # 1) Perform the eigen factorization of the matrix -> UVU'
    eigen_mat = eigen_factorize_cov_mat(
        HessInvProductCovarianceMatrix(hess_inv), n_pc, random_state
    )
    # 2) Convert U sqrt(V) to a sparse format
    sp_mat = sp.sparse.lil_array(eigen_mat.eig_vects * np.sqrt(eigen_mat.eig_vals).T)
    # 3) Resize -> we now have a square matrix and indices are preserved
    sp_mat.resize(eigen_mat.shape)
    # 4) Convert to column format
    return sp_mat.tocsc()


def factorize_hess_inv(
    hess_inv: LbfgsInvHessProduct,
    size_threshold: int = 200,
    n_pc: int = 200,
    random_state: Optional[
        Union[int, np.random.Generator, np.random.RandomState]
    ] = None,
) -> Union[csc_array, NDArrayFloat]:
    """
    Factorize the linear operator for the L-BFGS approximate inverse Hessian as AA^T.

    A is a sparse square matrix with size (n, n) with only the n_pc first columns
    non null.

    Parameters
    ----------
    hess_inv : LbfgsInvHessProduct
        Linear operator for the L-BFGS approximate inverse Hessian.
        This operator computes the product of a vector with the approximate inverse
        of the Hessian of the objective function, using the L-BFGS limited memory
        approximation to the inverse Hessian, accumulated during the optimization.
    size_threshold : int, optional
        If the number of elements n in the diagonal of hess_inv of size (n, n) is
        smaller or equal to the `size_threshold`, then a cholesky factorization is used
        and the lower factor is returned. Above this number, n is considered to large
        to perform cholesky and an eigen factorization is used instead. By default 200.
    n_pc : int
        Number of principal component kept (eigen vectors and values). The default is
        200.
    random_state : Optional[Union[int, np.random.Generator, np.random.RandomState]]
        Pseudorandom number generator state used to generate resamples.
        If `random_state` is ``None`` (or `np.random`), the
        `numpy.random.RandomState` singleton is used.
        If `random_state` is an int, a new ``RandomState`` instance is used,
        seeded with `random_state`.
        If `random_state` is already a ``Generator`` or ``RandomState``
        instance then that instance is used. The default is None

    Returns
    -------
    Union[csc_array, NDArrayFloat]
        Lower cholesky factor or sparse square factor A with size (n, n) and only the
        n_pc first columns non null.
    """
    # if the matrix is reasonably small, se perform a cholesky factorization directly
    # on the dense matrix
    if hess_inv.shape[0] <= size_threshold:
        return sp.linalg.cholesky(hess_inv.todense(), lower=True)

    # Otherwise, we use an eigen factorization.
    # we cannot have n_pc equal or higher than the number of adjusted parameters
    return get_sparse_eigen_factor_hess(
        hess_inv, max(1, min(hess_inv.shape[0] - 1, n_pc)), random_state
    )


def sample_from_hess_inv_post_dist(
    param: dminv.AdjustableParameter,
    x_cond: NDArrayFloat,
    hess_inv: LbfgsInvHessProduct,
    n_samples: int = 100,
    n_pc: int = 200,
    size_threshold: int = 200,
    random_state: Optional[
        Union[int, np.random.Generator, np.random.RandomState]
    ] = None,
) -> NDArrayFloat:
    """
    Sample from the Quantify the uncertainty from the ibverse lbfgsb hessian.

    If no preconditioning, then simply solve N_s symstem to extract the diagonal.
    Otherwise, perform a sparse cholesky factorization and generated samples that are
    backconditioned to propagate the uncertainty.

    Parameters
    ----------
    random_state : Optional[Union[int, np.random.Generator, np.random.RandomState]]
        Pseudorandom number generator state used to generate resamples.
        If `random_state` is ``None`` (or `np.random`), the
        `numpy.random.RandomState` singleton is used.
        If `random_state` is an int, a new ``RandomState`` instance is used,
        seeded with `random_state`.
        If `random_state` is already a ``Generator`` or ``RandomState``
        instance then that instance is used. The default is None
    """
    # Comme il est très difficile d'obtenir des forme analytiques de propagation
    # d'incertitudes, on procède de la manière suivante:
    # 1) Nous factorisons l'inverse du hessien (la matrice de covariance a posteriori)
    # sous la forme AA'
    # 2) Cette factorization nous permet de génerer des réalizations equiprobables des
    # paramètres préconditionés a posteriori, sachant que le vecteurs valeurs est la
    # moyenne.
    # 3) On approxime la variance à partir de l'ensemble de réalizations
    # back-conditionnées. Ces réalisations peuvent être utilisées pour propager les
    # incertitudes à travers le solveur direct.
    # uncertainty propagation
    factor = factorize_hess_inv(hess_inv, size_threshold, n_pc, random_state)

    if random_state is not None:
        random_state = check_random_state(random_state)
    else:
        random_state = np.random.default_rng()

    # ens = np.zeros((param.size, n_samples))
    # for i in range(n_samples):
    # Random state for v0 vector used by eigsh and svds

    return np.vstack(
        [
            [
                param.preconditioner.backtransform(factor @ w + x_cond)
                for w in random_state.normal(size=(n_samples, hess_inv.shape[0]))
            ]
        ]
    ).T

    # if param.preconditioner is NoTransform():
    #     # No need to transform the parameters, the diagonal can be extracted directly
    #     # TODO: if the transformation is linear, same thing.
    #     return extract_hess_inv_diag(hess_inv)
```
